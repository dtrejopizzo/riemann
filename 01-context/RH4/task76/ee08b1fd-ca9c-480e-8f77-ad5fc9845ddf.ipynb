```python
# Build chi mod 5: there are 4 characters mod 5. Non-principal ones include the real one (Legendre 5)
# and complex ones of order 4. "F2: L(s, chi_4), chi_4 mod 5" - chi_4 is the 4th character mod 5.
# The 4 chars mod 5: chi_0 (principal), chi_1, chi_2 (real, Legendre), chi_3 (complex conjugate of chi_1)
# The naming convention "chi_4" typically refers to the 4th in some indexing. 
# In LMFDB, characters mod 5 are labeled 5.a, 5.b (real Legendre), 5.c, 5.d (complex)
# The most common "chi_4 mod 5" refers to a character of order 4. Let's use the order-4 character.

# Order-4 char mod 5: chi(2) = i, chi(3) = -i, chi(4) = -1, chi(1) = 1, chi(5)=0
# This gives chi(2^k) cycling through i, -1, -i, 1 etc.
# Let's verify Davenport-Heilbronn: it's specifically defined with this complex character mod 5.
# The classical DH: L_DH(s) = L(s, chi) + (1-i*xi)/(1+i*xi) * L(s, chi-bar) -- no wait
# Actually DH uses real coefficients defined: a_n is one of {1, ξ, -ξ, -1, 0} for n mod 5.
# In Titchmarsh §10.25: f(s) = ((1-iκ)/2) L(s,χ) + ((1+iκ)/2) L(s, χ̄) where χ is order-4 mod 5
# This has real Dirichlet coefficients

# Let me look up: DH coefficients are real with pattern depending on n mod 5:
# c(n) = 1 if n ≡ 1 mod 5
# c(n) = κ if n ≡ 2 mod 5
# c(n) = -κ if n ≡ 3 mod 5
# c(n) = -1 if n ≡ 4 mod 5
# c(n) = 0 if n ≡ 0 mod 5
# This is the standard form. With κ = (sqrt(10-2sqrt(5))-2)/(sqrt(5)-1) ≈ 0.28408...

kappa = 0.28408 # validated value

def dh_coefs(N, kappa):
 a = np.zeros(N, dtype=complex)
 for i, n in enumerate(range(1, N+1)):
 r = n % 5
 if r == 1:
 a[i] = 1
 elif r == 2:
 a[i] = kappa
 elif r == 3:
 a[i] = -kappa
 elif r == 4:
 a[i] = -1
 # r==0 -> 0
 return a

a_F4 = dh_coefs(N, kappa)
a_F5p = dh_coefs(N, kappa * 1.05)
a_F5m = dh_coefs(N, kappa * 0.95)

# F2: L(s, chi_4) chi_4 mod 5 - the order-4 character (complex)
def chi4_mod5_coefs(N):
 # chi(1)=1, chi(2)=i, chi(3)=-i, chi(4)=-1, chi(0)=0
 table = {1:1, 2:1j, 3:-1j, 4:-1, 0:0}
 a = np.zeros(N, dtype=complex)
 for i, n in enumerate(range(1, N+1)):
 a[i] = table[n % 5]
 return a

a_F2 = chi4_mod5_coefs(N)

print("F4 first 10:", a_F4[:10].real)
print("F2 first 10:", a_F2[:10])
```

```python
# Good. Now let's set up the computation pipeline.
# 
# D(t; N) = sum_{n=1}^{N} a_n / sqrt(n) * exp(-i*t*log(n)) (analytic-norm Dirichlet polynomial)
# S_k(t) = sum_{n: omega(n)=k} a_n / sqrt(n) * exp(-i*t*log(n))
# 
# Total: D(t) = sum_k S_k(t)
# 
# Step 1: For each class, compute |D(t;999)|^2 on a fine grid in [10^4, 2e4]; find top 200 peaks
# Step 2: At those peak t values, compute S_k(t) for k=0..4 (5 strata)
# Step 3: M_jk = mean_t [Re(S_j(t) * conj(S_k(t)))] -> 5x5 real matrix
# Step 4: pc1_complex = lambda_1 / sum(eigenvalues), H_complex = -sum(p_i log p_i)/log(5) where p_i = lambda_i/sum

# Coefficient arrays (length 999, indexed [0..998] for n=1..999)
# Build them for all 12 classes

n_arr = np.arange(1, N+1) # length 999
inv_sqrt_n = 1.0 / np.sqrt(n_arr)
log_n = np.log(n_arr)
omega_arr = omega[1:N+1] # length 999, omega_arr[i] = omega(i+1)
# For F13, we use omega up to n=998; for n=999, set b=0 effectively
# omega_arr already has 999 values

# Build coefficient arrays a[n] for n=1..999
coeffs = {}
coeffs['F1'] = np.ones(N, dtype=complex)
coeffs['F2'] = a_F2
coeffs['F4'] = a_F4
coeffs['F5p'] = a_F5p
coeffs['F5m'] = a_F5m
coeffs['F6'] = a_F6
coeffs['F7'] = a_F7
coeffs['F9'] = a_F9
coeffs['F10'] = a_F10
coeffs['F11'] = a_F11
coeffs['F12'] = a_F12
# F13: a_F13 has a[0]=0, a[i]=a(n=i) for i=0..998. We need n=1..999.
# So coeffs['F13'][i] for i=0..998 corresponds to n=1..999.
# coeffs['F13'][0..997] = a_F13[1..998] (n=1..998), coeffs['F13'][998] = 0 (n=999, missing)
F13_coeffs = np.zeros(N, dtype=complex)
F13_coeffs[:998] = a_F13[1:999]
coeffs['F13'] = F13_coeffs

for k, v in coeffs.items():
 print(f"{k}: len={len(v)}, |a_1|={abs(v[0]):.4f}, |a_2|={abs(v[1]):.4f}")
```

```python
# Now compute D(t) = sum a_n/sqrt(n) * n^{-i*t} for t in dense grid
# We need fine grid to find peaks. Use t in [10^4, 2*10^4], grid spacing ~ 2*pi/log(N)/(oversampling)
# Effective resolution scale: dt ~ 2*pi/(log(N)) ~ 0.91
# Use dt = 0.05 for good peak resolution

# Build D(t) using vectorized computation: 
# D(t) = sum_n b_n * exp(-i*t*log(n)) where b_n = a_n / sqrt(n)
# For grid of T values, D = exp(-i * outer(T, log_n)) @ b

t_grid = np.arange(1e4, 2e4 + 0.05, 0.05)
print("t_grid size:", len(t_grid))
# For 200000 t-values * 999 coeffs = 2e8 complex operations per class. ~30 sec/class
# 12 classes => ~6 min. Doable.

# Better: do it in chunks to manage memory
def compute_D(coef, t_grid, log_n, inv_sqrt_n):
 b = coef * inv_sqrt_n # complex 999
 # Process in chunks
 chunk = 5000
 D = np.zeros(len(t_grid), dtype=complex)
 for i in range(0, len(t_grid), chunk):
 T = t_grid[i:i+chunk]
 # phase[i,j] = exp(-i * T[i] * log_n[j])
 phase = np.exp(-1j * np.outer(T, log_n))
 D[i:i+chunk] = phase @ b
 return D

# Test with F1 (zeta) - quick check
import time
t0 = time.time()
D_F1 = compute_D(coeffs['F1'], t_grid, log_n, inv_sqrt_n)
print(f"F1 D computed in {time.time()-t0:.1f}s, |D|_max = {np.abs(D_F1).max():.3f}")
```

```python
# Good, 8.6s/class * 12 = ~104s for D. Manageable.
# Now find top 200 peaks
from scipy.signal import find_peaks

absD_F1 = np.abs(D_F1)
peaks_idx, _ = find_peaks(absD_F1)
print(f"F1 number of peaks found: {len(peaks_idx)}")
peak_heights = absD_F1[peaks_idx]
top200_in_peaks = np.argsort(peak_heights)[-200:]
top_peak_idx = peaks_idx[top200_in_peaks]
top_peak_t = t_grid[top_peak_idx]
print(f"F1 top 5 peak heights: {sorted(peak_heights)[-5:]}")
print(f"F1 top 5 peak t: {sorted(top_peak_t)[-5:]}")
```

```python
# Now compute S_k(t) for k=0..4 at the 200 peaks
def compute_Sk_at_peaks(coef, t_peaks, log_n, inv_sqrt_n, omega_arr, k_max=4):
 b = coef * inv_sqrt_n
 n_peaks = len(t_peaks)
 S = np.zeros((k_max+1, n_peaks), dtype=complex)
 # For each k stratum, compute sum over n with omega(n)=k
 for k in range(k_max+1):
 mask = (omega_arr == k)
 if mask.sum() == 0:
 continue
 b_k = b[mask]
 log_n_k = log_n[mask]
 # phase[i,j] = exp(-i*t[i]*log(n_k[j])); 200 x ~500 -> small
 phase = np.exp(-1j * np.outer(t_peaks, log_n_k))
 S[k, :] = phase @ b_k
 return S # shape (5, n_peaks)

# Test with F1
S_F1 = compute_Sk_at_peaks(coeffs['F1'], top_peak_t, log_n, inv_sqrt_n, omega_arr, k_max=4)
print("S_F1 shape:", S_F1.shape)
# Verify D = sum_k S_k
D_check = S_F1.sum(axis=0)
D_actual = D_F1[top_peak_idx]
print(f"max |D_check - D_actual| = {np.abs(D_check - D_actual).max():.2e}")
```

```python
# Great, sanity check passes. Now compute features.
def spectral_features(S):
 # M_jk = E_t [Re(S_j(t) conj(S_k(t)))]
 # S shape (K, T)
 K, T = S.shape
 M = np.zeros((K, K))
 for j in range(K):
 for k in range(K):
 M[j, k] = np.mean((S[j] * np.conj(S[k])).real)
 # M is symmetric. Get eigenvalues
 eigvals = np.linalg.eigvalsh(M)
 # Sort descending. Note M may have negative eigenvalues if there's noise; but it's a Gram-like matrix and should be PSD up to mean operation.
 # Actually M_jk = mean[Re(S_j conj S_k)] = Re(mean[S_j conj S_k]) is the real part of a Hermitian matrix
 # The full Hermitian matrix M^H_jk = mean[S_j conj S_k] is PSD.
 # The real part isn't necessarily PSD. Let's just use absolute values for stability
 eigvals_pos = np.maximum(eigvals, 0) # clip tiny negatives
 eigvals_sorted = np.sort(eigvals_pos)[::-1]
 total = eigvals_sorted.sum()
 if total == 0:
 return np.nan, np.nan
 pc1 = eigvals_sorted[0] / total
 p = eigvals_sorted / total
 p = p[p > 0]
 H = -np.sum(p * np.log(p)) / np.log(K)
 return pc1, H

pc1, H = spectral_features(S_F1)
print(f"F1: pc1 = {pc1:.4f}, H = {H:.4f}")
print(f"Reference (saved feature_pc1[0], feature_H[0]): {pc1_existing[0]:.4f}, {H_existing[0]:.4f}" if False else "")
print(f"Saved F1 from feature_pc1.npy: {np.load('feature_pc1.npy')[0]:.4f}, feature_H.npy: {np.load('feature_H.npy')[0]:.4f}")
```

```python
# pc1 matches very closely (0.9563 vs 0.9565), but H differs (0.1284 vs 0.2056)
# Likely the entropy normalization differs. Let me try: 
# - log(K=5) vs log(num_nonzero)?
# - Maybe they don't clip negatives

def spectral_features_v2(S, clip_neg=True):
 K, T = S.shape
 M = np.zeros((K, K))
 for j in range(K):
 for k in range(K):
 M[j, k] = np.mean((S[j] * np.conj(S[k])).real)
 eigvals = np.linalg.eigvalsh(M)
 if clip_neg:
 eigvals = np.maximum(eigvals, 0)
 eigvals_sorted = np.sort(np.abs(eigvals))[::-1] # use absolute values?
 total = eigvals_sorted.sum()
 pc1 = eigvals_sorted[0] / total
 p = eigvals_sorted / total
 p = p[p > 0]
 H_log_n = -np.sum(p * np.log(p)) / np.log(len(p)) if len(p) > 1 else 0
 H_log_K = -np.sum(p * np.log(p)) / np.log(K)
 H_natural = -np.sum(p * np.log(p))
 return pc1, H_log_K, H_log_n, H_natural, eigvals_sorted

pc1, Hk, Hn, Hnat, evals = spectral_features_v2(S_F1, clip_neg=False)
print(f"F1 no clip: pc1={pc1:.4f}, H/log5={Hk:.4f}, H/log(non-zero)={Hn:.4f}, H_natural={Hnat:.4f}")
print(f" eigenvalues: {evals}")
# Saved F1 H = 0.2056. Try with abs(eigvals):
```

```python
# H_natural = 0.2066 is very close to saved 0.2056. 
# So H_complex = -sum p log(p) WITHOUT normalization by log(K). Use natural log.

def spectral_features_v3(S):
 K, T = S.shape
 M = np.zeros((K, K))
 for j in range(K):
 for k in range(K):
 M[j, k] = np.mean((S[j] * np.conj(S[k])).real)
 eigvals = np.linalg.eigvalsh(M)
 eigvals = np.maximum(eigvals, 0) # clip negatives (numerical noise)
 total = eigvals.sum()
 if total == 0:
 return np.nan, np.nan
 p = eigvals / total
 pc1 = p.max()
 p_pos = p[p > 1e-15]
 H = -np.sum(p_pos * np.log(p_pos))
 return pc1, H

pc1, H = spectral_features_v3(S_F1)
print(f"F1: pc1 = {pc1:.4f}, H = {H:.4f}")
print(f"Saved: pc1 = {np.load('feature_pc1.npy')[0]:.4f}, H = {np.load('feature_H.npy')[0]:.4f}")
```

```python
# Excellent! Match within rounding. The tiny diff (0.9563 vs 0.9565, 0.2066 vs 0.2056) is likely from
# different t_grid resolution / peak-finding details. Acceptable.
# 
# Now, let's run the full pipeline for all 12 classes
import time

def find_top_peaks(D_abs, t_grid, n_top=200):
 peaks_idx, _ = find_peaks(D_abs)
 heights = D_abs[peaks_idx]
 top_in_peaks = np.argsort(heights)[-n_top:]
 top_idx = peaks_idx[top_in_peaks]
 return top_idx, t_grid[top_idx], heights[top_in_peaks]

# Run for all classes
class_names = ['F1', 'F2', 'F4', 'F5p', 'F5m', 'F6', 'F7', 'F9', 'F10', 'F11', 'F12']
results = {}

for cls in class_names:
 t0 = time.time()
 D = compute_D(coeffs[cls], t_grid, log_n, inv_sqrt_n)
 absD = np.abs(D)
 _, peak_t, peak_h = find_top_peaks(absD, t_grid, 200)
 S = compute_Sk_at_peaks(coeffs[cls], peak_t, log_n, inv_sqrt_n, omega_arr, k_max=4)
 pc1, H = spectral_features_v3(S)
 elapsed = time.time() - t0
 results[cls] = {'pc1': pc1, 'H': H, 'peak_t': peak_t, 'peak_h': peak_h, 'S': S}
 print(f"{cls}: pc1={pc1:.4f}, H={H:.4f}, max_peak={peak_h.max():.3f} ({elapsed:.1f}s)")
```
