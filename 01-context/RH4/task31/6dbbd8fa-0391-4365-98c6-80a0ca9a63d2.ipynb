```python
# So pre-computed class-level features exist for F1, F2, F6, F7, F9, F13, F14, F15, F16. 
# Missing: F4, F5p, F5m, F10, F11, F12. We can compute these from existing per-peak S_k arrays.
#
# Define key functions:
import numpy as np

def compute_M_pc1_H(S):
 """
 S: (n_peaks, 8) complex array of stratified partial sums S_k(t_p), k=0..7
 Returns:
 M (8x8 complex Hermitian): M_jk = E_p [ S_j(t_p) * conj(S_k(t_p)) ]
 pc1_complex: largest eigenvalue / sum of eigenvalues (concentration in PC1)
 H_complex: spectral entropy (-sum p_i log p_i) / log(8)
 """
 n, k = S.shape
 # peak-averaged Gram matrix
 M = (S.conj().T @ S) / n # M_{jk} = (1/n) sum_p conj(S_j) S_k -- note ordering
 # We want M_jk = E[S_j conj(S_k)]. Use S.T @ S.conj() / n
 M = (S.T @ S.conj()) / n
 # Hermitian eigenvalues
 eigs = np.linalg.eigvalsh((M + M.conj().T)/2) # ensure Hermitian
 eigs = np.sort(eigs)[::-1]
 eigs = np.maximum(eigs, 0)
 s = eigs.sum()
 if s == 0:
 return M, np.nan, np.nan
 p = eigs / s
 pc1 = p[0]
 # spectral entropy normalized
 nz = p[p>0]
 H = -(nz * np.log(nz)).sum() / np.log(len(p))
 return M, pc1, H

# Validate against pre-computed F13 features
S_F13 = np.load('S_F13.npy')
M, pc1, H = compute_M_pc1_H(S_F13)
print(f'F13: pc1={pc1:.6f}, H={H:.6f}')
print(f'Reference: pc1={float(np.load("pc1_complex_F13.npy")):.6f}, H={float(np.load("H_complex_F13.npy")):.6f}')
```

```python
# omega_array only has 1000 entries (for F13's small-N analysis). We need omega(n) for n=1..10^6 to use F14's coeffs.
# Let me generate that.

import numpy as np

def small_omega_sieve(N):
 """Number of distinct prime factors omega(n) for n in 1..N."""
 omega = np.zeros(N+1, dtype=np.int8)
 is_composite = np.zeros(N+1, dtype=bool)
 for p in range(2, N+1):
 if not is_composite[p]:
 # p is prime
 omega[p::p] += 1
 is_composite[p*p::p] = True
 return omega

# This is slow in pure Python at N=10^6. Vectorize:
def small_omega_sieve_fast(N):
 omega = np.zeros(N+1, dtype=np.int8)
 sieve = np.zeros(N+1, dtype=bool) # is_prime
 sieve[2:] = True
 for p in range(2, N+1):
 if sieve[p]:
 omega[p::p] += 1
 if p*p <= N:
 sieve[p*p::p] = False
 return omega

import time
t0 = time.time()
omega_full = small_omega_sieve_fast(10**6)
print(f'Computed omega for N=10^6 in {time.time()-t0:.2f}s')
print('omega range:', omega_full.min(), omega_full.max())
print('omega(1)=', omega_full[1], 'omega(2)=', omega_full[2], 'omega(6)=', omega_full[6], 'omega(30)=', omega_full[30])
```

```python
# Now define stratified S_k(t) computation:
# S_k(t) = sum_{n=1..N, omega(n)=k} a_n / n^{1/2} * n^{-it}
# We use the binding rule R3 normalization: lambda_n = a_n / n^{(weight-1)/2} for L-functions of weight w
# But for our peak-feature computation, the existing code seems to already use raw a_n values.
# 
# Actually looking at the data: F14 has a_n_real/imag stored. Standard "Dirichlet partial sum" is 
# S(t) = sum_n a_n n^{-1/2} cos(t log n) -- analytic-normalized so a_n already includes the 1/n^{(w-1)/2} factor.
# 
# But how was S computed for F1-F12 originally? Let's check: F1.npy at peak F1_t[0]:
# S_F1[0] should be sum_{omega(n)=k} 1 * n^{-1/2-i*t_0} for k=0..7
# 
# Let's verify by computing S for F1 at first peak using a_n=1 (all ones, like Riemann zeta):
import numpy as np
from numba import njit, prange

@njit(parallel=True, fastmath=False, cache=True)
def stratified_sk(a_n, omega, t, omega_max):
 """
 Compute S_k(t) = sum_{n: omega(n)=k} a_n * n^{-1/2 - i*t}, using Kahan summation.
 a_n: (N,) complex array (1-indexed shifted to 0-indexed, n=i+1)
 omega: (N,) int array
 t: scalar
 omega_max: max k value (returns array length omega_max+1)
 """
 N = a_n.shape[0]
 K = omega_max + 1
 # Per-stratum running sums and Kahan compensators (real/imag separately)
 sum_re = np.zeros(K)
 sum_im = np.zeros(K)
 c_re = np.zeros(K)
 c_im = np.zeros(K)
 for n_idx in range(N):
 n = n_idx + 1
 k = omega[n_idx]
 if k >= K:
 continue
 # term = a_n[n_idx] * n^{-1/2} * exp(-i*t*log(n))
 log_n = np.log(n)
 amp = 1.0 / np.sqrt(n)
 ang = -t * log_n
 cos_a = np.cos(ang)
 sin_a = np.sin(ang)
 ar = a_n[n_idx].real
 ai = a_n[n_idx].imag
 term_re = (ar*cos_a - ai*sin_a) * amp
 term_im = (ar*sin_a + ai*cos_a) * amp
 # Kahan
 y_re = term_re - c_re[k]
 t_re = sum_re[k] + y_re
 c_re[k] = (t_re - sum_re[k]) - y_re
 sum_re[k] = t_re
 y_im = term_im - c_im[k]
 t_im = sum_im[k] + y_im
 c_im[k] = (t_im - sum_im[k]) - y_im
 sum_im[k] = t_im
 out = np.empty(K, dtype=np.complex128)
 for k in range(K):
 out[k] = sum_re[k] + 1j*sum_im[k]
 return out

# Quick test: compute S for F1 (a_n=1) at first peak t and compare with stored F1.npy[0]
N = 10**6
omega_full = small_omega_sieve_fast(N)
omega_arr = omega_full[1:N+1].astype(np.int64)
a_F1 = np.ones(N, dtype=np.complex128)
t_F1_first = float(np.load('F1_t.npy')[0])
print(f'Computing S_k for F1 at t={t_F1_first}...')
import time
t0=time.time()
S_test = stratified_sk(a_F1, omega_arr, t_F1_first, 7)
print(f'Time: {time.time()-t0:.2f}s')
print('Computed S_F1[0] (k=0..7):', S_test)
print('Stored F1[0] (k=0..7):', np.load('F1.npy')[0])
```

```python
# Perfect match! The computation reproduces stored F1 data exactly.
# Now compute S_k for F14 at its 200 peaks using its stored coefficients.
import numpy as np
import time

a_F14_re = np.load('a_n_real.npy')
a_F14_im = np.load('a_n_imag.npy')
a_F14 = a_F14_re + 1j*a_F14_im
t_F14 = np.load('selected_t.npy')

print('F14: a_n shape:', a_F14.shape, 'peaks:', t_F14.shape, 't range:', t_F14.min(), t_F14.max())

# Compute S_k for all 200 peaks
t0 = time.time()
S_F14 = np.empty((len(t_F14), 8), dtype=np.complex128)
for i, t in enumerate(t_F14):
 S_F14[i] = stratified_sk(a_F14, omega_arr, float(t), 7)
 if i % 50 == 0:
 print(f' peak {i}/{len(t_F14)} t={t:.0f} elapsed={time.time()-t0:.1f}s')
print(f'F14 total time: {time.time()-t0:.1f}s')
print('S_F14 first row:', S_F14[0])

# Save
np.save('F14_Sk_recomputed.npy', S_F14)

# Compute features
M, pc1, H = compute_M_pc1_H(S_F14)
print(f'F14 recomputed: pc1={pc1:.6f}, H={H:.6f}')
print(f'F14 reference : pc1={float(np.load("pc1_complex_F14.npy")):.6f}, H={float(np.load("H_complex_F14.npy")):.6f}')
```

```python
# Train OC-SVM. Use a moderate gamma and nu.
from sklearn.svm import OneClassSVM
from sklearn.metrics import roc_auc_score

# Try a few hyperparameters and pick reasonable defaults
oc = OneClassSVM(kernel='rbf', gamma='scale', nu=0.05).fit(X_train_s)

# Decision function: positive = inlier, negative = outlier. Higher = more in-distribution.
# For AUC, we treat "anomaly score" = -decision_function so higher means more anomalous.

# Sanity: training inlier rate
train_pred = oc.predict(X_train_s) # +1 inlier, -1 outlier
print(f'Training inlier rate: {(train_pred==1).mean():.3f} (expected ~{1-0.05:.2f})')

# Now for each violator family compute AUC vs the training-set GRH-true peaks
# The objective requires AUC > 0.9 across all violator families. We compute AUC where:
# positive class = "violator" (anomaly)
# negative class = "GRH-true" (use training set inliers, OR held-out GRH-true)
# Standard practice: combine GRH-true holdout (or training as proxy) vs each violator family.

# Build evaluation function
def eval_family(name, scaler, oc, df_peaks, features, true_classes, violator_class):
 df_true = df_peaks[df_peaks['class'].isin(true_classes)]
 df_viol = df_peaks[df_peaks['class']==violator_class]
 X_true = scaler.transform(df_true[features].values)
 X_viol = scaler.transform(df_viol[features].values)
 score_true = -oc.decision_function(X_true) # higher = more anomalous
 score_viol = -oc.decision_function(X_viol)
 y = np.concatenate([np.zeros(len(score_true)), np.ones(len(score_viol))])
 s = np.concatenate([score_true, score_viol])
 auc = roc_auc_score(y, s)
 pred_viol = oc.predict(X_viol)
 flag_rate = (pred_viol==-1).mean()
 return auc, flag_rate, len(score_true), len(score_viol)

# Evaluate against training GRH-true (in-sample reference) and held-out F6/F7 (out-of-sample)
print('\nEvaluation: training-set GRH-true (F1,F2,F9,F10,F11,F13) as the negative class')
print(f'{"family":<10}{"AUC":>10}{"flag_rate":>15}{"n_true":>10}{"n_viol":>10}')
for v in violators:
 auc, flag, n_true, n_viol = eval_family(v, scaler, oc, df_peaks, features, grh_true_train_classes, v)
 print(f'{v:<10}{auc:>10.4f}{flag:>15.3f}{n_true:>10}{n_viol:>10}')

print('\nEvaluation: held-out GRH-true (F6,F7) as the negative class')
print(f'{"family":<10}{"AUC":>10}{"flag_rate":>15}{"n_true":>10}{"n_viol":>10}')
for v in violators:
 auc, flag, n_true, n_viol = eval_family(v, scaler, oc, df_peaks, features, grh_true_holdout_classes, v)
 print(f'{v:<10}{auc:>10.4f}{flag:>15.3f}{n_true:>10}{n_viol:>10}')
```

```python
# Important observations:
# - Against training-set GRH-true negatives, OC-SVM achieves AUC > 0.99 on F4/F5p/F5m/F12 and 0.898 on F14.
# - Against held-out GRH-true (F6 Liouville, F7 Möbius), AUC drops dramatically because F6/F7 themselves get
# scored as anomalies by an OC-SVM trained on F1/F2/F9/F10/F11/F13. This reveals F6/F7 are outliers vs the
# training distribution. 
# - F14 AUC is below 0.9 in both cases. F14 used full N=10^6 with t in [10^4, 2*10^4] — protocol mismatch with
# training-set classes that have peaks mostly in t > 22000.
# 
# Hypothesis is "AUC > 0.9 across all violator families". With the in-sample reference, 4/5 violator families 
# meet AUC > 0.9 (F4, F5p, F5m, F12); F14 misses at 0.898. With held-out GRH-true negatives, only F12 meets the
# threshold — so the hypothesis is NOT supported under a strict held-out evaluation.
# 
# Let's also examine F14 more carefully. The training set includes F13 which uses N=999. Training set is mixed-N.
# Try training without F13 and see if F14 changes.

X_train2 = df_peaks[df_peaks['class'].isin(['F1','F2','F9','F10','F11'])][features].values
scaler2 = StandardScaler().fit(X_train2)
X_train2_s = scaler2.transform(X_train2)
oc2 = OneClassSVM(kernel='rbf', gamma='scale', nu=0.05).fit(X_train2_s)
print('Trained on F1,F2,F9,F10,F11 (excl F13).')
print(f'\n{"family":<10}{"AUC":>10}{"flag_rate":>12}')
for v in violators:
 auc, flag, _, _ = eval_family(v, scaler2, oc2, df_peaks, features, ['F1','F2','F9','F10','F11'], v)
 print(f'{v:<10}{auc:>10.4f}{flag:>12.3f}')
```

```python
# F6 (Liouville) and F7 (Möbius) themselves get flagged as anomalies and often score MORE anomalously than 
# violators. This drags AUC down sharply. F6/F7 use sign-only multiplicative coefficients — ±1, no growth — 
# which differs from the modular L-functions in the training set.
# 
# Bootstrap CIs to be statistically rigorous (per binding rule R5/R6 mentioned in memory).
import numpy as np
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score

rng = np.random.default_rng(42)
n_boot = 200

def bootstrap_aucs(scaler, oc, df_peaks, features, true_classes, violator):
 df_true = df_peaks[df_peaks['class'].isin(true_classes)]
 df_viol = df_peaks[df_peaks['class']==violator]
 s_true = -oc.decision_function(scaler.transform(df_true[features].values))
 s_viol = -oc.decision_function(scaler.transform(df_viol[features].values))
 aucs = []
 for _ in range(n_boot):
 i = rng.integers(0, len(s_true), len(s_true))
 j = rng.integers(0, len(s_viol), len(s_viol))
 y = np.concatenate([np.zeros(len(i)), np.ones(len(j))])
 s = np.concatenate([s_true[i], s_viol[j]])
 aucs.append(roc_auc_score(y, s))
 return np.array(aucs)

print('Bootstrap AUC 95% CIs')
print(f'\nTrain={["F1","F2","F9","F10","F11"]}; vs in-sample GRH-true negatives')
print(f'{"family":<8}{"AUC":>9}{"95% CI":>22}')
for v in violators:
 aucs = bootstrap_aucs(scaler2, oc2, df_peaks, features, ['F1','F2','F9','F10','F11'], v)
 print(f'{v:<8}{aucs.mean():>9.4f} [{np.quantile(aucs,0.025):.4f}, {np.quantile(aucs,0.975):.4f}]')

print(f'\nTrain={["F1","F2","F9","F10","F11"]}; vs held-out GRH-true F6,F7')
print(f'{"family":<8}{"AUC":>9}{"95% CI":>22}')
for v in violators:
 aucs = bootstrap_aucs(scaler2, oc2, df_peaks, features, ['F6','F7'], v)
 print(f'{v:<8}{aucs.mean():>9.4f} [{np.quantile(aucs,0.025):.4f}, {np.quantile(aucs,0.975):.4f}]')
```
