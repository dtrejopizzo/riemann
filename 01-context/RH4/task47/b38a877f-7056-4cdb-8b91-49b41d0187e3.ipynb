```python

# Plan:
# 1. Compute D_N = sum_{n<=N} n^{-s} for ζ on the critical line (Re s = 1/2). 
# We need to choose t for the partial sum. Since the metric R_comp is defined
# "for ζ at N=10^6", but the problem doesn't specify a specific t, the most
# natural averaged formulation is via mean values. However, to keep this Step
# 0.4 controlled experiment clean, we will follow the most standard approach:
# evaluate R_comp at s=1/2 (i.e. on the real axis at the critical line's
# real part) where the partial sum diverges -- problematic.
#
# Actually, with the ambiguity, the cleanest approach: R_comp at s=1/2+i*0 isn't
# meaningful. The canonical "completeness/coherence ratio" for ω-stratified
# Dirichlet partial sums in -style analysis typically refers to a moment
# decomposition. Without a specified t, the natural surrogate is the *average*
# behavior. 
# 
# Given the hypothesis specifies ~30% contribution -- and Sathe-Selberg gives
# semiprime density at log log N ≈ 2.6 for N=10^6, the k=2 cohort is a major
# contributor.
#
# Per the task: we will define R_comp as the (incoherent / coherent) second-
# moment ratio averaged over a range of heights t along the critical line.
# That is, for ζ:
# D_N(t) = Σ_{n≤N} n^{-1/2-it} (Kahan-summed)
# S_k(t) = Σ_{n≤N, ω(n)=k} n^{-1/2-it} 
# E[X] = average over t in some sample
# R_comp = E[Σ_k |S_k|^2] / E[|D_N|^2] (incoherent fraction; ≥1, with 1 iff cross terms cancel; this matches M_coh = 1/R_comp interpretation)
#
# However at N=10^6 this is an enormous computation. Let's first assess feasibility.
# 
# Actually the most computationally realistic and theoretically sound approach
# given the resources: compute |S_k|^2 / |D_N|^2 at a single (or several) heights
# t away from any zero of ζ, where all sums converge to meaningful values.
#
# Let me first do a feasibility check.

print("Planning computation...")
```

```python

# Step 1: Compute ω(n) (number of distinct prime factors) for all n up to N=10^6
# using a linear sieve. This is feasible.
import numpy as np
import time

def compute_omega(N):
 """Compute omega(n) (number of distinct prime factors) for n=0..N."""
 omega = np.zeros(N+1, dtype=np.int8)
 for p in range(2, N+1):
 if omega[p] == 0 or p == 2: # p might be prime if first time set
 pass
 # Better: standard linear sieve approach
 omega = np.zeros(N+1, dtype=np.int8)
 for p in range(2, N+1):
 if omega[p] == 0: # p is prime (omega still 0)
 for k in range(p, N+1, p):
 omega[k] += 1
 return omega

t0 = time.time()
N_test = 10**5 # Quick test
omega_test = compute_omega(N_test)
print(f"omega for N=10^5 took {time.time()-t0:.2f}s")
print(f"Max omega at N=10^5: {omega_test[1:].max()}")
# Distribution
unique, counts = np.unique(omega_test, return_counts=True)
print("Distribution:", dict(zip(unique.tolist(), counts.tolist())))
```

```python

# Good. Now compute for N=10^6.
t0 = time.time()
N = 10**6
omega_full = compute_omega(N)
print(f"omega for N=10^6 took {time.time()-t0:.2f}s")
print(f"Max omega at N=10^6: {omega_full[1:].max()}")
unique, counts = np.unique(omega_full, return_counts=True)
print("Distribution:", dict(zip(unique.tolist(), counts.tolist())))
# 510510 = 2*3*5*7*11*13*17 ≤ 10^6, so max ω = 7
print(f"omega(510510) = {omega_full[510510]}")
```

```python

# Good. ω(1)=0, max ω = 7 at N=10^6. There are 8 "ω=0" actually 1 from n=1 plus n=0 (we ignore 0).
# omega[0] = 0 by initialization but is not a real number. Distribution shows 0:2 but that includes n=0.
# omega[1] = 0 (correct). So at N=10^6 there are k=0..7 strata.
print(f"omega[0]={omega_full[0]}, omega[1]={omega_full[1]}") # both 0 in array

# Now define R_comp.
# Given the ambiguity, I'll adopt the most defensible operational definition that
# is consistent with the document: the "incoherence completeness ratio"
# R_comp = Σ_k |S_k|^2 / |D_N|^2
# and M_coh = 1/R_comp = |D_N|^2 / Σ_k |S_k|^2 = "coherent gain" (≤1, =1 iff ω-strata add coherently with no cancellation).
# 
# Note this is invariant under multiplication of all S_k by a common phase.
# However at s on the real line for ζ partial sums, |D_N|^2 is huge, and all S_k are
# positive (real), so R_comp would essentially equal Σ|S_k|^2/(Σ|S_k|)^2.
# This is the classic "participation ratio" and is BETWEEN 1/K and 1. With K=8 strata
# and most weight on k=2,3, this would be near 1/3 or so. Let's check on real line first.

# Compute partial sums at s = 1/2 (real). Use Kahan summation per R6.
def kahan_sum(arr):
 s = 0.0
 c = 0.0
 for x in arr:
 y = x - c
 t = s + y
 c = (t - s) - y
 s = t
 return s

# Vectorized Kahan via numpy is non-trivial, but we can use a simple compensated sum.
# For our 10^6 floats this won't be a major issue with float64. Use numpy.fsum (uses Shewchuk algorithm; arbitrary precision sum).
# math.fsum is exact-rounded. Use that.
import math

n_arr = np.arange(1, N+1, dtype=np.float64)
a_n = n_arr ** (-0.5) # zeta coefficients: ζ partial sum on real axis at s=1/2
# Total D_N
D_N = math.fsum(a_n)
print(f"D_N at s=1/2 (real) = {D_N:.6f}")
print(f"Naive: {a_n.sum():.6f}")

# Now per-stratum sums S_k = Σ_{ω(n)=k} a_n (for n≥1; ignore n=0)
omega_n = omega_full[1:N+1] # omega(1)..omega(N)
S_k = {}
for k in range(0, 8):
 mask = (omega_n == k)
 if mask.any():
 S_k[k] = math.fsum(a_n[mask])
 else:
 S_k[k] = 0.0
print("S_k at s=1/2 (real):")
for k, v in S_k.items():
 print(f" k={k}: S_k = {v:.4f}, count = {(omega_n == k).sum()}")
print(f"Σ S_k = {sum(S_k.values()):.6f}, D_N = {D_N:.6f}")
```

```python

# Good check: Σ S_k = D_N to full precision.
# Now on the real axis, all S_k are real and positive, so |D_N|^2 and Σ |S_k|^2:
sum_sq_S = sum(v**2 for v in S_k.values())
D_N_sq = D_N**2
R_comp_real = sum_sq_S / D_N_sq
print(f"On real axis at s=1/2 (degenerate, no oscillation):")
print(f" Σ |S_k|^2 = {sum_sq_S:.4f}")
print(f" |D_N|^2 = {D_N_sq:.4f}")
print(f" R_comp (incoh fraction) = {R_comp_real:.6f}")
print(f" M_coh = 1/R_comp = {1/R_comp_real:.6f}")
# This is just the participation ratio and shows S_3 dominates here, not S_2.
# This is not really meaningful since at s=1/2 (real) the partial sum is just sum of positives.
# We need to evaluate at s = 1/2 + i*t for some t (critical line).

# Per document, the relevant statistic for ζ at N=10^6 is presumably evaluated
# along the critical line. Given the hypothesis specifies a deterministic ~30%
# contribution, it's most likely an average over a range of t. Let me use the
# averaged second-moment approach.
#
# By orthogonality (mean value theorem): for T → ∞, 
# (1/T) ∫_0^T (m/n)^{it} dt → 1 if m=n, else 0.
# Therefore: E_t |D_N(1/2+it)|^2 = Σ_n 1/n + corrections,
# and similarly E_t |S_k|^2 = Σ_{ω(n)=k} 1/n.
# Cross terms E[S_j conj(S_k)] = Σ_{ω(m)=j, ω(n)=k, m=n} 1/m = 0 for j≠k.
# So in this asymptotic-averaged sense, Σ_k |S_k|^2 = |D_N|^2 exactly (up to MVT corrections).
# So averaged over large T, R_comp = 1. This is not useful.
#
# Alternative: the key quantity at finite t may be the SECOND moment expectation:
# E[|D_N|^2] = Σ_n 1/n
# E[|S_k|^2] = Σ_{ω(n)=k} 1/n
# Then "R_comp" could be defined as the ratio 
# R_comp = E[|S_2|^2] / E[|D_N|^2] ≈ Σ_{ω(n)=2} 1/n / Σ_n 1/n
# But that's just one stratum's fraction.
#
# A more general R_comp matching the "completeness" name:
# R_comp = Σ_k E[|S_k|^2] / E[|D_N|^2]
# This is identically 1 by MVT — not useful either.

# Let me reconsider: the key insight is that R_comp must be computable as a single
# scalar, and "zeroing S_k" must change it. Therefore R_comp must depend on
# something other than Σ|S_k|^2 (which trivially decomposes).
#
# A natural candidate that gives ω-dependent contributions when we zero an S_k:
# R_comp = |D_N|^2 / Σ_n |a_n|^2 (compression ratio)
# = (Σ_n a_n)^2 / Σ_n a_n^2 (for real a_n). Setting S_k=0 means zeroing those a_n.
# Then |D_N|^2 changes (squared sum drops), and so does Σ|a_n|^2.
#
# But more interestingly, the standard "R_comp" in compressive sensing / sparse
# representations is:
# R_comp = (Σ |a_n|)^2 / (N · Σ |a_n|^2) (or similar)
# = inverse participation ratio.
# 
# Without document specification I'll proceed with a clean, theoretically motivated
# definition that matches all known constraints, and TRANSPARENTLY state this.

print("\nProceeding with computational experiment using a defensible operational definition.")
```

```python

# IMPORTANT OBSERVATION: at s=1/2 on the real axis, R_comp = Σ|S_k|^2/|D_N|^2 = 0.2829, 
# which is REMARKABLY close to ~30%! And the largest stratum contribution by |S_k|^2/|D_N|^2 is...

contributions = {k: v**2 / D_N_sq for k, v in S_k.items()}
print("|S_k|^2 / |D_N|^2 contributions:")
for k, c in contributions.items():
 print(f" k={k}: {c:.6f} = {100*c:.3f}%")
print(f"Sum: {sum(contributions.values()):.6f}")
# k=3 dominates here, not k=2.
# But the metric R_comp = 0.2829 = Σ|S_k|^2/|D_N|^2.
# When we zero S_2 (set those S to 0), we recompute D_N and R_comp.

# Test: let's run the controlled experiment.
def compute_R_comp(S_k_dict):
 """R_comp = Σ|S_k|^2 / |Σ S_k|^2"""
 D = sum(S_k_dict.values())
 sumsq = sum(v**2 for v in S_k_dict.values())
 if D == 0:
 return float('nan')
 return sumsq / (D**2)

R_full = compute_R_comp(S_k)
print(f"\nR_comp (full) = {R_full:.6f}")
for k_zero in [1, 2, 3, 4]:
 S_mod = {k: (0.0 if k==k_zero else v) for k, v in S_k.items()}
 R_mod = compute_R_comp(S_mod)
 delta = R_mod - R_full
 rel = delta / R_full * 100
 print(f"Zeroing S_{k_zero}: R_comp = {R_mod:.6f}, ΔR = {delta:+.6f}, rel change = {rel:+.2f}%")
```

```python

# Important: this real-axis computation is degenerate (no oscillatory cancellation) and
# the results don't show a unique ω=2 effect — all of S_1..S_4 cause comparable ~20-40% changes.
# The hypothesis claim that ω=2 specifically drives ~30% with others having NO effect 
# is NOT supported on the real axis. 
#
# But the document explicitly says the metric should be evaluated for ζ - which lives on
# complex line. Let me compute using actual complex partial sums at t > 0 (critical line).
# 
# I'll compute |D_N(1/2+it)|^2 and |S_k(1/2+it)|^2 for many random t and average.
# This is the ω-stratified second moment.
#
# Computational consideration: for one t, computing N=10^6 terms of n^{-1/2-it} is fast
# in numpy (vectorized). For multiple t's we can compute many (~hundreds).

def compute_complex_partial_sums(N, t_vals, omega_arr):
 """For each t, compute D_N(1/2+it) and S_k(1/2+it) for k=0..7. 
 Returns dict keyed by k, value is array over t."""
 n = np.arange(1, N+1, dtype=np.float64)
 log_n = np.log(n)
 sqrt_n = np.sqrt(n)
 omega_n = omega_arr[1:N+1]
 
 n_t = len(t_vals)
 K = 8
 S_kt = np.zeros((K, n_t), dtype=np.complex128)
 D_t = np.zeros(n_t, dtype=np.complex128)
 
 # masks per k
 masks = [omega_n == k for k in range(K)]
 
 for i, t in enumerate(t_vals):
 # a_n = n^{-1/2 - i t} = (1/sqrt(n)) * exp(-i t log n)
 phase = np.exp(-1j * t * log_n)
 a_n = phase / sqrt_n
 D_t[i] = a_n.sum() # using numpy sum; could use math.fsum but float128 not avail
 for k in range(K):
 S_kt[k, i] = a_n[masks[k]].sum()
 return D_t, S_kt

# Test with a few t values first
t_test = np.array([14.134725, 21.022040, 25.010858, 30.424876]) # near ζ zeros
D_t, S_kt = compute_complex_partial_sums(N, t_test, omega_full)
print("At ζ zeros (small t):")
for i, t in enumerate(t_test):
 D = D_t[i]
 Sk = S_kt[:, i]
 print(f"\nt={t}: |D_N|={abs(D):.4f}, |D_N|^2={abs(D)**2:.4f}")
 print(f" Σ|S_k|^2 = {(np.abs(Sk)**2).sum():.4f}")
 print(f" R_comp = {(np.abs(Sk)**2).sum() / abs(D)**2:.4f}")
 for k in range(8):
 if abs(Sk[k]) > 1e-10:
 print(f" k={k}: |S_k|^2 = {abs(Sk[k])**2:.4f}, frac = {abs(Sk[k])**2/abs(D)**2:.4f}")
```

```python

# Good. R_comp ≈ 0.28-0.30 across these test points, robust at ~30%.
# Now run the controlled experiment with many random t values.
# 
# Strategy: 
# 1. Sample 200 random t values uniformly in [10, 1000] (well above the
# real-axis singularity, in the regime where MVT averaging applies meaningfully).
# 2. For each t, compute D_N, S_k.
# 3. Compute baseline R_comp = mean over t of (Σ|S_k|^2 / |D_N|^2).
# 4. For each k_zero in {1,2,3,4}: zero those a_n contributions, recompute
# D_N' = Σ S_k (k≠k_zero), recompute R_comp' = mean over t of (Σ_{k≠k_zero}|S_k|^2 / |D_N'|^2).
# 5. Effect size: (R_comp - R_comp_baseline) per stratum-zeroing, with
# block bootstrap CI over t-samples.

# Time per t: ~few seconds for N=10^6 with 8 mask sums. Let's optimize.

def compute_complex_partial_sums_fast(N, t_vals, omega_arr, log_n=None, sqrt_n_inv=None):
 """Optimized version. Returns S_kt[K, n_t] array with k=0..7 strata; D_t is sum over k."""
 if log_n is None:
 n = np.arange(1, N+1, dtype=np.float64)
 log_n = np.log(n)
 sqrt_n_inv = 1.0 / np.sqrt(n)
 
 omega_n = omega_arr[1:N+1]
 K = 8
 n_t = len(t_vals)
 S_kt = np.zeros((K, n_t), dtype=np.complex128)
 
 # Pre-compute masks 
 masks = [omega_n == k for k in range(K)]
 
 # For each k separately: only multiply phase for those n's
 for k in range(K):
 mask = masks[k]
 if not mask.any():
 continue
 log_n_k = log_n[mask]
 sqrt_n_inv_k = sqrt_n_inv[mask]
 # S_k(t) = Σ sqrt_n_inv * exp(-i*t*log_n)
 # vectorize: shape (len(mask_k), n_t)
 # For memory: split into batches if needed
 for i, t in enumerate(t_vals):
 phase = np.exp(-1j * t * log_n_k)
 S_kt[k, i] = (sqrt_n_inv_k * phase).sum()
 return S_kt

import time
n_arr_full = np.arange(1, N+1, dtype=np.float64)
log_n_full = np.log(n_arr_full)
sqrt_n_inv_full = 1.0 / np.sqrt(n_arr_full)

# Test timing for 50 t values
np.random.seed(42)
t_sample = np.random.uniform(10, 1000, 50)
t0 = time.time()
S_kt = compute_complex_partial_sums_fast(N, t_sample, omega_full, log_n_full, sqrt_n_inv_full)
print(f"50 t-values for N=10^6 took {time.time()-t0:.1f}s")
```

```python

# Excellent — fast enough. Let's run with 200 t values for robust averaging.
np.random.seed(20260509)
n_t = 200
# Sample t in a moderate range — large enough for MVT, modest enough that partial
# sums at N=10^6 still capture full series well (t << N).
t_sample = np.random.uniform(50, 5000, n_t)
t0 = time.time()
S_kt = compute_complex_partial_sums_fast(N, t_sample, omega_full, log_n_full, sqrt_n_inv_full)
print(f"200 t-values for N=10^6 took {time.time()-t0:.1f}s")

# D_t = sum over k of S_kt
D_t = S_kt.sum(axis=0)
abs_D2 = np.abs(D_t)**2
abs_S2 = np.abs(S_kt)**2 # shape (K, n_t)

# Per-t R_comp
R_comp_t = abs_S2.sum(axis=0) / abs_D2
print(f"R_comp per t: mean={R_comp_t.mean():.6f}, std={R_comp_t.std():.6f}")
print(f" min={R_comp_t.min():.4f}, max={R_comp_t.max():.4f}")

# Baseline: mean
R_baseline = R_comp_t.mean()
print(f"\nBaseline R_comp (averaged over t): {R_baseline:.6f}")
```

```python

# At large t values, individual |D_N|^2 fluctuates wildly, with mean << individual peaks.
# This is the regime where ζ partial sums oscillate. The distinction between 
# "ratio of means" vs "mean of ratios" is critical.
#
# By MVT, for ζ on the critical line, mean_t |D_N(1/2+it)|^2 ≈ Σ_n 1/n ≈ log N + γ ≈ 14.39 for N=10^6.
# Let's verify:
print(f"Σ 1/n for n=1..10^6 = {math.fsum(1/n for n in range(1, N+1)):.4f}")
print(f"mean |D_N(1/2+it)|^2 over our 200 t-samples = {abs_D2.mean():.4f}")
# These should match to leading order. Let me check why the mean is much smaller.
# The mean over 200 random t in [50, 5000] should converge to Σ 1/n ≈ 14.39, but
# we see 7.58. Hmm, surely with 200 samples there's variance, but median R_comp_t = 3.5
# suggests at typical t we get small |D_N|^2 (sums oscillate a lot).
# The MVT result is for T → ∞ asymptotically. At finite t with N=10^6 partial sums,
# there's a square-root cancellation regime where |D_N|^2 ~ Σ 1/n.

# Let's use much higher t range (where MVT is a better approximation).
# Actually at higher t, the partial sum oscillates more. The MVT integrates over
# many periods of all (m/n)^{it} terms.
# 
# Actually the MVT is: integral 0 to T of |D_N(1/2+it)|^2 dt = T*Σ|a_n|^2/n + 2*Re Σ_{m<n} ... 
# = T*(Σ 1/n) + O(N^2). So for T much larger than N^2, the mean is Σ 1/n.
# Our T-window length is 4950 << N^2 = 10^12, so MVT not yet applicable.
# 
# Let's check Σ 1/n vs Σ_{m<n} (m n)^{-1/2} * sin(...)
# 
# We need to use the appropriate regime. Conventionally, "metric for ζ at N=10^6"
# without specifying t -- and given the deliverable is a Jacobian ∂R_comp/∂S_k --
# I should evaluate R_comp at SPECIFIC, reproducible t values (not averaged).
# 
# Best choice: evaluate at the canonical low ζ zeros (where D_N is small) AND at
# typical points (where it is large). Or simply at t=1 (s = 1/2 + i, well-defined).
# 
# Given the document does not specify t explicitly, the cleanest reproducible choice is:
# evaluate at a fixed reasonable t (e.g., t=1), and PRIMARILY the metric should be
# computed in the asymptotic-MVT sense where it has theoretical interpretation.

# Let me use the EXPECTATION-BASED definition:
# R_comp = (Σ_n 1/n with ω(n)>0) / (Σ_n 1/n) -- this is the "weight" of each ω-stratum
# in the second-moment expectation.
# This is exactly:
# E[|S_k|^2] / E[|D_N|^2] in the MVT-asymptotic sense (cross terms vanish).
# And R_comp = Σ_k E[|S_k|^2] / E[|D_N|^2] = 1 trivially (sum to 1). So not useful as a single scalar.
#
# But the real quantity of interest in Step 0.4 is: "what's the contribution 
# of each ω-stratum to E[|D_N|^2]?". The answer is just E[|S_k|^2]/E[|D_N|^2].
# Let's compute that — it has a clean theoretical answer.

# Theoretical: E[|S_k|^2] = Σ_{ω(n)=k} 1/n (exact MVT result for n^{-1/2}; cross terms vanish)
# E[|D_N|^2] = Σ_{n} 1/n
# fraction_k = Σ_{ω(n)=k} 1/n / Σ_n 1/n

theoretical_E_Sk_sq = {}
omega_n = omega_full[1:N+1]
n_arr = np.arange(1, N+1, dtype=np.float64)
inv_n = 1.0 / n_arr
total_E = math.fsum(inv_n.tolist())
print(f"E[|D_N|^2] = Σ 1/n = {total_E:.6f}")
for k in range(8):
 mask = (omega_n == k)
 if mask.any():
 theoretical_E_Sk_sq[k] = math.fsum(inv_n[mask].tolist())
 else:
 theoretical_E_Sk_sq[k] = 0.0
 print(f" k={k}: E[|S_k|^2] = {theoretical_E_Sk_sq[k]:.6f}, fraction = {theoretical_E_Sk_sq[k]/total_E:.6f} = {100*theoretical_E_Sk_sq[k]/total_E:.3f}%")
```

```python

# Under Definition I (theoretical E[|D_N|^2]):
# Zero S_2: -35.6% (close to 30%, but not exactly)
# Zero S_1: -25.4% 
# Zero S_3: -23.6%
# Zero S_4: -7.4%
# So ω=2 is the LARGEST contributor at ~36%, but ω=1 (25%) and ω=3 (24%) are also
# substantial contributors. The hypothesis "30% specific to ω=2" is PARTIALLY supported
# (ω=2 is dominant) but the additional claim "not present for ω=1, 3, or 4" is REFUTED.

# Now let's add bootstrap CIs for the empirical version under Definition (b)
# (since Definition I is theoretical with no sampling variance).

# Block bootstrap on the t-samples for definition (b):
def bootstrap_def_b(S_kt, n_boot=2000, rng=None):
 """Bootstrap-resample t-samples to compute CIs for ratio-of-means R_comp."""
 if rng is None:
 rng = np.random.default_rng(20260509)
 n_t = S_kt.shape[1]
 abs_S2 = np.abs(S_kt)**2 # (K, n_t)
 sum_Sk_sq = abs_S2.sum(axis=0) # (n_t,)
 abs_D2 = np.abs(S_kt.sum(axis=0))**2 # (n_t,)
 
 # Pre-compute zeroed versions
 keys = ['full'] + [f'zero_{k}' for k in range(1,5)]
 nums = {'full': sum_Sk_sq}
 dens = {'full': abs_D2}
 for k_zero in range(1, 5):
 S_mod = S_kt.copy()
 S_mod[k_zero, :] = 0
 nums[f'zero_{k_zero}'] = np.abs(S_mod)**2 .sum(axis=0)
 dens[f'zero_{k_zero}'] = np.abs(S_mod.sum(axis=0))**2
 
 boot_results = {key: np.zeros(n_boot) for key in keys}
 for b in range(n_boot):
 idx = rng.integers(0, n_t, n_t)
 for key in keys:
 boot_results[key][b] = nums[key][idx].mean() / dens[key][idx].mean()
 return boot_results

print("Running bootstrap...")
t0 = time.time()
boot_b = bootstrap_def_b(S_kt, n_boot=2000)
print(f"Bootstrap took {time.time()-t0:.1f}s")

# Report
print("\nDefinition (b) — R_comp = mean(Σ|S_k|^2) / mean(|D_N|^2)")
print("="*75)
R_full = boot_b['full'].mean()
R_full_ci = np.percentile(boot_b['full'], [2.5, 97.5])
print(f" Full: {R_full:.6f} [95% CI: {R_full_ci[0]:.6f}, {R_full_ci[1]:.6f}]")
for k_zero in [1, 2, 3, 4]:
 R_z = boot_b[f'zero_{k_zero}'].mean()
 delta_boot = boot_b[f'zero_{k_zero}'] - boot_b['full']
 delta_mean = delta_boot.mean()
 delta_ci = np.percentile(delta_boot, [2.5, 97.5])
 rel_boot = 100 * delta_boot / boot_b['full']
 rel_mean = rel_boot.mean()
 rel_ci = np.percentile(rel_boot, [2.5, 97.5])
 print(f" Zero S_{k_zero}: R={R_z:.4f}, ΔR={delta_mean:+.4f} CI[{delta_ci[0]:+.4f}, {delta_ci[1]:+.4f}], "
 f"rel={rel_mean:+.2f}% CI[{rel_ci[0]:+.2f}%, {rel_ci[1]:+.2f}%]")
```

```python

# Under empirical Definition (b) with bootstrap:
# - Zeroing S_1 has small effect (+1.7%, CI crosses 0)
# - Zeroing S_2: +7.5%, CI [-8.6%, +21.3%] (crosses 0)
# - Zeroing S_3: +8.5%, CI [-13.0%, +29.6%] (crosses 0)
# - Zeroing S_4: +7.5%, CI [-9.7%, +25.4%] (crosses 0)
# So under (b), NO ω stratum produces a significant ~30% effect.
#
# The hypothesis is REFUTED under the empirical ratio-of-means definition (b).
# Under the theoretical (asymptotic MVT) Definition I, ω=2 contributes 35.6% but
# ω=1 (25.4%) and ω=3 (23.6%) also contribute substantially -- so "specific to ω=2"
# is FALSE under that definition either.

# Let me also run the analysis under Definition I (theoretical) since it's clean.
# Bootstrap there is unnecessary (it's exact arithmetic on Σ 1/n).
# But I can compute confidence intervals based on the empirical sample-moment
# estimator: estimate E[|S_k|^2] from observed |S_k(t_i)|^2 mean.

# Definition I_emp: R_comp_I_emp = mean over t of |D_N(t)|^2 ≈ Σ 1/n
# Zero S_k: R_comp = mean over t of |D_N - S_k|^2 (with renormalization)
# Or: R_comp = mean over t of Σ_{j≠k} |S_j|^2 (stratum k removed; DON'T renorm D_N)
# Latter is cleaner.

def bootstrap_def_I_emp(S_kt, n_boot=2000, rng=None):
 """R_comp = E[|D_N|^2] estimated from samples; zeroing S_k means computing 
 R_comp_zeroed = E[|D_N - S_k|^2]"""
 if rng is None:
 rng = np.random.default_rng(20260509)
 n_t = S_kt.shape[1]
 abs_D2_full = np.abs(S_kt.sum(axis=0))**2 # |D_N|^2
 
 keys = ['full'] + [f'zero_{k}' for k in range(1,5)]
 abs_D2 = {'full': abs_D2_full}
 for k_zero in range(1,5):
 S_mod = S_kt.copy()
 S_mod[k_zero, :] = 0
 abs_D2[f'zero_{k_zero}'] = np.abs(S_mod.sum(axis=0))**2
 
 boot = {k: np.zeros(n_boot) for k in keys}
 for b in range(n_boot):
 idx = rng.integers(0, n_t, n_t)
 for k in keys:
 boot[k][b] = abs_D2[k][idx].mean()
 return boot

boot_I = bootstrap_def_I_emp(S_kt, n_boot=2000)
print("Definition I (E[|D_N|^2] estimator): R_comp = mean over t of |D_N|^2")
print("="*75)
R_full = boot_I['full'].mean()
R_full_ci = np.percentile(boot_I['full'], [2.5, 97.5])
print(f" Full: R_comp = {R_full:.4f} [95% CI: {R_full_ci[0]:.4f}, {R_full_ci[1]:.4f}]")
print(f" (Theoretical Σ 1/n = {total_E:.4f})")
for k_zero in [1, 2, 3, 4]:
 R_z = boot_I[f'zero_{k_zero}'].mean()
 R_z_ci = np.percentile(boot_I[f'zero_{k_zero}'], [2.5, 97.5])
 delta_boot = boot_I[f'zero_{k_zero}'] - boot_I['full']
 delta_mean = delta_boot.mean()
 delta_ci = np.percentile(delta_boot, [2.5, 97.5])
 rel_boot = 100 * delta_boot / boot_I['full']
 rel_mean = rel_boot.mean()
 rel_ci = np.percentile(rel_boot, [2.5, 97.5])
 print(f" Zero S_{k_zero}: R={R_z:.4f} CI[{R_z_ci[0]:.4f}, {R_z_ci[1]:.4f}], "
 f"ΔR={delta_mean:+.4f} CI[{delta_ci[0]:+.4f}, {delta_ci[1]:+.4f}], "
 f"rel={rel_mean:+.2f}% CI[{rel_ci[0]:+.2f}%, {rel_ci[1]:+.2f}%]")
```

```python

# Under Definition I (E[|D_N|^2]):
# Zero S_1: -27.4% [CI: -37%, -19%]
# Zero S_2: -37.5% [CI: -43%, -28%] <- close to "30%" claim
# Zero S_3: -28.2% [CI: -47%, -7%] <- ALSO close to 30%!
# Zero S_4: -13.3% [CI: -29%, +3%] <- smaller and CI crosses 0
#
# So under Definition I:
# - ω=2 effect is ~37% (close to 30% claim, perhaps within tolerance)
# - ω=1 effect is ~27% (also close to 30%!)
# - ω=3 effect is ~28% (also close to 30%!)
# - ω=4 effect is ~13% (smaller, CI may cross 0)
# 
# Multiple strata (ω=1, 2, 3) all produce ~25-37% effects. ω=2 is largest but NOT
# uniquely so. The hypothesis "30% effect specific to ω=2" is REFUTED by the 
# observation that ω=1 and ω=3 also produce comparable effects.

# Final note: bootstrap CIs for Δ_2 vs Δ_1 and Δ_3 — do they overlap? 
# Test pairwise:
delta_1 = boot_I['zero_1'] - boot_I['full']
delta_2 = boot_I['zero_2'] - boot_I['full']
delta_3 = boot_I['zero_3'] - boot_I['full']
delta_4 = boot_I['zero_4'] - boot_I['full']

# Difference distributions (same bootstrap indices used so paired)
# We can compute relative: the size of |Δ_2| should be much larger than |Δ_1|, |Δ_3|, |Δ_4|
diff_21 = delta_2 - delta_1 # if neg, S_2 effect is larger (more negative)
diff_23 = delta_2 - delta_3
diff_24 = delta_2 - delta_4
print("Pairwise differences in Δ (zero_k vs zero_2): negative => S_2 effect larger")
for name, arr in [('Δ_2 - Δ_1', diff_21), ('Δ_2 - Δ_3', diff_23), ('Δ_2 - Δ_4', diff_24)]:
 m = arr.mean()
 ci = np.percentile(arr, [2.5, 97.5])
 print(f" {name}: mean={m:+.4f}, 95% CI=[{ci[0]:+.4f}, {ci[1]:+.4f}], P(<0)={(arr<0).mean():.3f}")
```

```python

# Pairwise comparisons under Definition I:
# - S_2 effect vs S_1 effect: P(S_2 effect larger) = 76%; CI overlaps zero — NOT distinguishable
# - S_2 effect vs S_3 effect: P(S_2 effect larger) = 82%; CI overlaps zero — NOT distinguishable
# - S_2 effect vs S_4 effect: P(S_2 effect larger) = 100%; CI [-5.29, -2.47] excludes 0 — DISTINGUISHABLE
# 
# Conclusion: ω=2 effect is statistically distinguishable from ω=4 (which is half),
# but NOT distinguishable from ω=1 or ω=3.
# The hypothesis "ω=2 uniquely responsible for ~30% change" is REFUTED.

# Now create the final summary figure.
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

# Panel A: theoretical fractional contributions of each ω-stratum to E[|D_N|^2]
ax = axes[0]
ks = list(range(8))
fractions = [theoretical_E_Sk_sq[k]/total_E for k in ks]
colors = ['gray']*8
colors[2] = 'red' # highlight ω=2
bars = ax.bar(ks, [100*f for f in fractions], color=colors, edgecolor='black')
ax.axhline(30, color='red', linestyle='--', alpha=0.6, label='30% reference')
ax.set_xlabel('ω(n) — number of distinct prime factors')
ax.set_ylabel('Theoretical contribution to E[|D_N|²] (%)')
ax.set_title('A. ω-stratified contributions to second moment of ζ partial sum (N=10⁶)')
ax.legend()
for k, f in zip(ks, fractions):
 ax.text(k, 100*f + 0.5, f'{100*f:.1f}%', ha='center', fontsize=9)
ax.set_ylim(0, 45)

# Panel B: empirical effect of zeroing each S_k (ΔR_comp/R_comp) with bootstrap CIs
ax = axes[1]
ks_test = [1, 2, 3, 4]
mean_rel = []
ci_lo = []
ci_hi = []
for k in ks_test:
 delta = boot_I[f'zero_{k}'] - boot_I['full']
 rel = 100 * delta / boot_I['full']
 mean_rel.append(rel.mean())
 ci = np.percentile(rel, [2.5, 97.5])
 ci_lo.append(rel.mean() - ci[0])
 ci_hi.append(ci[1] - rel.mean())
colors2 = ['gray', 'red', 'gray', 'gray']
ax.errorbar(ks_test, mean_rel, yerr=[ci_lo, ci_hi], fmt='o', capsize=8, 
 markersize=12, color='black', ecolor='black', linewidth=2)
for k, m, c in zip(ks_test, mean_rel, colors2):
 ax.scatter([k], [m], color=c, s=200, zorder=5, edgecolor='black', linewidth=1.5)
ax.axhline(-30, color='red', linestyle='--', alpha=0.6, label='Hypothesized -30% effect')
ax.axhline(0, color='black', linestyle='-', alpha=0.3)
ax.set_xlabel('ω(n) of stratum set to zero')
ax.set_ylabel('ΔR_comp / R_comp (%)')
ax.set_title('B. Effect of zeroing each S_k on R_comp = E[|D_N|²], with 95% bootstrap CIs')
ax.legend()
ax.set_xticks(ks_test)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('R_comp_omega_decomposition.png', dpi=300, bbox_inches='tight')
plt.savefig('R_comp_omega_decomposition.pdf', bbox_inches='tight')
plt.show()
print("Saved figure: R_comp_omega_decomposition.png and .pdf")
```
