```python
# Now we have cypari2 ready. The script's resume logic depends on CHECKPOINT files in checkpoints_ldh/
# Looking at the script: it scans ldh_ckpt_*.npy and resumes from latest checkpoint, not from ldh_zeros_partial_dps50.npy
# We need to seed a checkpoint from the partial file (2772 zeros). We have only ldh_zeros_partial_dps50.npy with 2772, 
# but the description says 3549 - that file is NOT available; we must work with the 2772 we have.
# We need to create a fake checkpoint named e.g. ldh_ckpt_02750.npy with first 2750 zeros, and meta with t_last.

# However the partial save has zeros up to index 2772 (last t≈2827.7). The script's checkpoint loading reads only n_zeros from arr length, not meta.
# Let's create a checkpoint file that matches the partial save, with corresponding meta JSON containing t_last.

import os, json, numpy as np
WORK = '/workspace/ad42cbf2-61ea-4f79-8818-d835ae68a4d0'
CKPT_DIR = os.path.join(WORK, 'checkpoints_ldh')
os.makedirs(CKPT_DIR, exist_ok=True)

arr = np.load(os.path.join(WORK, 'ldh_zeros_partial_dps50.npy'), allow_pickle=True)
N = len(arr)
print('Partial length:', N)
# round down to nearest 250 -> 2750
n_ckpt = (N // 250) * 250 # 2750
print('Will create checkpoint at n =', n_ckpt)
label = f"{n_ckpt:05d}"
ckpt_arr = arr[:n_ckpt]
np.save(os.path.join(CKPT_DIR, f'ldh_ckpt_{label}.npy'), ckpt_arr, allow_pickle=True)
# t_last needs to be just below zero #n_ckpt (i.e. the last we processed). Use the actual zero value at index n_ckpt-1.
import mpmath
mpmath.mp.dps = 50
t_last_val = mpmath.mpf(str(ckpt_arr[-1]))
meta = {'n_zeros': n_ckpt, 't_last': mpmath.nstr(t_last_val, 30), 'n_evals': 0}
with open(os.path.join(CKPT_DIR, f'ldh_ckpt_{label}_meta.json'), 'w') as f:
 json.dump(meta, f)
print('Checkpoint written. t_last =', meta['t_last'])
# But this means script will resume at t = t_last (last zero), and the first sign change beyond will be skipped (because zeros[-1]=t_last_val).
# This is a problem because the loaded zeros include only first 2750. We have 2772 zeros, but only first 2750 included. The script will re-detect zeros 2751..2772 (22 zeros). Those will be added again as new findings - that's fine.
# Actually the script's resume only reads checkpoint files, ignoring the standalone partial file. So our partial standalone file is irrelevant for resume. The script-found zeros 2751..2772 will be re-found - cost is small.
```

```python
# 3465 zeros (added 693 zeros in ~24 min of compute). NOT 5000.
# Final figure summarizing the progress and projected runtime.
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import re

# Parse log file for PROGRESS entries with timestamps -> rate analysis
with open(os.path.join(WORK,'run_log.txt')) as f:
 log_lines = f.readlines()

# Extract zero count and t from PROGRESS lines (using this session's most recent run only, after 14:17 resume)
records = []
session_started = False
for line in log_lines:
 if '14:17:54' in line and 'Starting' in line:
 session_started = True
 continue
 if not session_started:
 continue
 m = re.search(r'\[(\d\d:\d\d:\d\d)\]\s+(?:PROGRESS|Checkpoint \d+):.*?(\d+) zeros.*?t=([\d\.]+).*?elapsed=([\d\.]+)min', line)
 if m:
 ts, nz, tval, el = m.groups()
 records.append((float(el), int(nz), float(tval)))

print(f'{len(records)} progress records this run')
print('Sample:', records[:3], records[-3:])

# Compute instantaneous z/min rate
import numpy as np
els = np.array([r[0] for r in records])
nzs = np.array([r[1] for r in records])
ts = np.array([r[2] for r in records])
# Rate per minute (forward differences)
rates = np.diff(nzs) / np.diff(els)
ts_mid = (ts[1:] + ts[:-1])/2

fig, axes = plt.subplots(2, 1, figsize=(8, 8))
ax = axes[0]
ax.plot(ts, nzs, 'b.-')
ax.axhline(5000, color='red', linestyle='--', label='Target (5000)')
ax.axhline(3465, color='green', linestyle=':', label=f'Achieved ({nzs[-1]})')
ax.set_xlabel('Imaginary height t')
ax.set_ylabel('Cumulative L_DH zeros found')
ax.set_title('A. L_DH zero generation progress (this session)')
ax.legend()
ax.grid(alpha=0.3)
ax.text(0.02, 0.95, 'Started: 3000 zeros (resumed from ckpt)\nEnded: 3465 zeros\nAdded: 465 zeros in ~23.7 min',
 transform=ax.transAxes, va='top', fontsize=9, bbox=dict(facecolor='white', alpha=0.8))

ax = axes[1]
ax.plot(ts_mid, rates, 'r.-')
# Project: linear fit
fit = np.polyfit(ts_mid, rates, 1)
t_proj = np.linspace(ts_mid[0], 5200, 100)
ax.plot(t_proj, np.polyval(fit, t_proj), 'k--', alpha=0.5, label=f'linear fit: rate≈{fit[0]:.4f}·t+{fit[1]:.2f}')
ax.set_xlabel('Imaginary height t')
ax.set_ylabel('Zeros per minute')
ax.set_title('B. Throughput declines with t — full 5000-zero target requires ≫1 h additional compute')
ax.grid(alpha=0.3)
ax.legend()
# Estimate remaining time to 5000
remaining = 5000 - nzs[-1]
avg_recent_rate = rates[-10:].mean() if len(rates) >= 10 else rates.mean()
est_min = remaining / avg_recent_rate
ax.text(0.02, 0.95, f'Avg recent rate: {avg_recent_rate:.1f} z/min\n'
 f'Remaining: {remaining} zeros\nEst. additional runtime: ≈{est_min:.0f} min ({est_min/60:.1f} h)',
 transform=ax.transAxes, va='top', fontsize=9, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
out_fig = os.path.join(WORK, 'ldh_generation_progress.png')
plt.savefig(out_fig, dpi=120, bbox_inches='tight')
print('Saved figure:', out_fig)
print(f'\nFinal result: {nzs[-1]} zeros (target 5000). Need {remaining} more, est ~{est_min:.0f} min additional compute.')
```
