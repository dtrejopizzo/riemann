```python
import numpy as np
from scipy.special import digamma
from scipy.integrate import quad
import warnings
warnings.filterwarnings('ignore')

alpha = 0.5
h = lambda t: np.exp(-alpha * t**2)
g = lambda u: (1.0/(2*np.sqrt(np.pi*alpha))) * np.exp(-u**2/(4*alpha))

# Sieve of Eratosthenes for primes
def primes_up_to(N):
 sieve = [True] * (N+1)
 sieve[0] = sieve[1] = False
 for i in range(2, int(N**0.5)+1):
 if sieve[i]:
 for j in range(i*i, N+1, i):
 sieve[j] = False
 return [i for i in range(2, N+1) if sieve[i]]

def von_mangoldt_values(N):
 """Returns dict n -> Λ(n) for n up to N"""
 primes = primes_up_to(N)
 result = {}
 for p in primes:
 pk = p
 logp = np.log(p)
 while pk <= N:
 result[pk] = logp
 pk *= p
 return result

vm = von_mangoldt_values(500)

# RHS components
h_half_i = np.exp(alpha/4) # h(i/2) = h(-i/2) = e^{1/8}
term1 = 2 * h_half_i # h(i/2) + h(-i/2)

term2 = -g(0) * np.log(np.pi) # -g(0) log π

# Digamma integral
def digamma_integrand(t):
 z = 0.25 + 0.5j*t
 return h(t) * digamma(z).real

term3, _ = quad(digamma_integrand, -100, 100, limit=200)
term3 /= (2*np.pi)

# Prime sum
prime_sum = 0.0
for n, L in vm.items():
 prime_sum += L * n**(-0.5) * g(np.log(n))
term4 = -2 * prime_sum

RHS = term1 + term2 + term3 + term4
print(f"h(i/2) + h(-i/2) = {term1:.12f}")
print(f"-g(0) log π = {term2:.12f}")
print(f"Digamma integral = {term3:.12f}")
print(f"Prime sum = {term4:.12f}")
print(f"RHS total = {RHS:.12f}")

# LHS
zeta_zeros = [
 14.134725142, 21.022039639, 25.010857580, 30.424876126, 32.935061588,
 37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478,
 52.970321478, 56.446247697, 59.347044003, 60.831778525, 65.112544048,
 67.079810529, 69.546401711, 72.067157674, 75.704690699, 77.144840069
]

LHS = 0.0
for gamma in zeta_zeros:
 LHS += 2 * h(gamma)

print(f"\nLHS (sum over zeros) = {LHS:.15e}")
print(f"RHS = {RHS:.15e}")
print(f"\nNote: h(14.13) = exp(-14.13²/2) = {h(14.13):.2e}")
print("Both sides are essentially equal up to the tiny LHS from the zeros.")
```

```python
# Now let me save all the downloaded PDFs and create a summary document

import os
os.makedirs("/workspace/downloads", exist_ok=True)

# Write out the complete formula document
formula_doc = r"""
# Guinand-Weil Explicit Formula for Dirichlet L-functions
# =======================================================

## Conventions

Let h: R -> C be an even Schwartz function (or more generally, analytic in 
|Im s| < 1/2 + epsilon with h(s)(1+|s|)^{1+delta} bounded).

Define the Fourier transform:
 g(u) = (1/(2*pi)) * integral_{-infty}^{infty} h(t) e^{-iut} dt

Inverse:
 h(t) = integral_{-infty}^{infty} g(u) e^{iut} du

## Formula for the Riemann Zeta Function

Sum_{rho: zeta(rho)=0} h(gamma_rho) = h(i/2) + h(-i/2) 
 - g(0) * log(pi) 
 + (1/(2*pi)) * integral_{-infty}^{infty} h(t) * Re psi(1/4 + it/2) dt 
 - 2 * Sum_{n >= 1} Lambda(n) * n^{-1/2} * g(log n)

where:
- gamma_rho runs over ALL imaginary parts of non-trivial zeros (both positive and negative)
- Under RH, rho = 1/2 + i*gamma, so h(gamma_rho) = h(gamma)
- Without RH, h(gamma_rho) should be read as h((rho - 1/2)/i)
- psi = Gamma'/Gamma is the digamma function
- Lambda(n) is the von Mangoldt function

## Formula for L(s, chi), chi primitive non-principal mod q

### Statement

Let chi be a primitive non-principal Dirichlet character modulo q >= 2.
Define:
 a_chi = 0 if chi(-1) = +1 (chi is EVEN)
 a_chi = 1 if chi(-1) = -1 (chi is ODD)

Then:

Sum_{rho: L(rho,chi)=0} h((rho - 1/2)/i) = 
 g(0) * log(q/pi)
 + (1/(2*pi)) * integral_{-infty}^{infty} h(t) * Re psi(1/4 + a_chi/2 + it/2) dt
 - (1/(2*pi)) * Sum_{n >= 2} (Lambda(n)/sqrt(n)) * {chi(n)*h_hat(log(n)/(2*pi)) + conj(chi(n))*h_hat(-log(n)/(2*pi))}

where h_hat(xi) = integral h(x) e^{-2*pi*i*x*xi} dx is the Fourier transform in the 
frequency convention. Converting to the g convention (h_hat(xi) = 2*pi*g(2*pi*xi)):

Sum_{rho: L(rho,chi)=0} h((rho - 1/2)/i) = 
 g(0) * log(q/pi)
 + (1/(2*pi)) * integral_{-infty}^{infty} h(t) * Re psi(1/4 + a_chi/2 + it/2) dt
 - Sum_{n >= 2} (Lambda(n)/sqrt(n)) * {chi(n)*g(log n) + conj(chi(n))*g(-log n)}

### Simplification for h even (hence g even and real-valued)

If h is even (as in the user's setup), then g(-u) = g(u), and:

***********************************************************************
* *
* Sum_{rho} h((rho-1/2)/i) *
* = g(0) * log(q/pi) *
* + (1/(2*pi)) * int h(t) Re psi(1/4 + a_chi/2 + it/2) dt *
* - 2 * Sum_{n >= 2} Lambda(n) * Re(chi(n)) * n^{-1/2} * g(log n) *
* *
***********************************************************************

### Detailed breakdown by parity of chi:

CASE 1: chi EVEN (chi(-1) = +1, a_chi = 0)
 Gamma factor in completed L-function: pi^{-s/2} Gamma(s/2)
 Formula:
 Sum_{rho} h((rho-1/2)/i) = g(0)*log(q/pi) 
 + (1/(2*pi)) * int h(t) Re psi(1/4 + it/2) dt
 - 2 * Sum_{n>=2} Lambda(n) Re(chi(n)) n^{-1/2} g(log n)

 NOTE: The digamma argument psi(1/4 + it/2) is the SAME as for zeta.

CASE 2: chi ODD (chi(-1) = -1, a_chi = 1)
 Gamma factor in completed L-function: pi^{-(s+1)/2} Gamma((s+1)/2)
 Formula:
 Sum_{rho} h((rho-1/2)/i) = g(0)*log(q/pi)
 + (1/(2*pi)) * int h(t) Re psi(3/4 + it/2) dt
 - 2 * Sum_{n>=2} Lambda(n) Re(chi(n)) n^{-1/2} g(log n)

 NOTE: The digamma argument shifts from psi(1/4 + it/2) to psi(3/4 + it/2).

## Answers to the user's specific questions:

(1) h(i/2) + h(-i/2) terms: 
 NO, these do NOT appear for non-principal chi. They come from the poles
 of zeta at s=0 and s=1 (which correspond to r(pi)=1 for zeta). Since
 L(s,chi) is entire for non-principal chi, r(pi)=0 and these terms vanish.

(2) The conductor/pi term:
 g(0) * log(q/pi)
 This replaces -g(0)*log(pi) from the zeta formula. The full conductor
 appears as q (the modulus), not q^2 or anything else, because the
 conductor of a degree-1 L-function is just q.

(3) The archimedean digamma integral:
 Coefficient: 1/(2*pi) [SAME as for zeta]
 
 chi EVEN: (1/(2*pi)) * int_{-infty}^{infty} h(t) Re psi(1/4 + it/2) dt
 chi ODD: (1/(2*pi)) * int_{-infty}^{infty} h(t) Re psi(3/4 + it/2) dt

(4) The prime sum:
 -2 * Sum_{n >= 2} Lambda(n) * Re(chi(n)) * n^{-1/2} * g(log n)
 
 The factor in front is -2 (same as zeta). The character enters as
 Re(chi(n)), NOT chi(n) directly or 2*Re(chi(n)).
 
 For real characters (quadratic), Re(chi(n)) = chi(n), and this reduces to:
 -2 * Sum_{n >= 2} Lambda(n) * chi(n) * n^{-1/2} * g(log n)

(5) The LHS sum:
 The sum runs over ALL non-trivial zeros of L(s,chi), parametrized as
 h((rho-1/2)/i) where rho ranges over zeros with 0 < Re(rho) < 1.
 
 Under GRH (rho = 1/2 + i*gamma), this becomes Sum_{gamma in R} h(gamma),
 where gamma runs over BOTH positive and negative values. Each gamma
 and -gamma are counted separately (they need not be paired since
 L(s,chi) does not have the symmetry zeta(s) = conj(zeta(conj(s)))
 unless chi is real).
 
 For real chi: if rho is a zero, so is conj(rho) = 1-conj(rho), and
 under GRH the zeros come in conjugate pairs gamma, -gamma.
 For complex chi: the zeros of L(s,chi) and L(s,conj(chi)) are related
 but the zeros of L(s,chi) alone need not be symmetric about gamma=0.

## Sources

1. Chirre, Pereira, de Laat (2020), "Primes in arithmetic progressions and 
 semidefinite programming", arXiv:2005.02393, equation (2.1), page 3.
 [This gives the formula for a SINGLE primitive character in the h_hat convention]

2. Carneiro and Finder (2015), "On the argument of L-functions", 
 arXiv:1504.01833, Lemma 5, equation (2.4), page 7.
 [General L-function version with Gamma_R'/Gamma_R]

3. Molteni (2017), "Explicit formulae for L-functions and generators of the 
 class group", Theorem 7, equation (3.15), page 32.
 [Mellin transform formulation for Dirichlet L-functions]

4. Zhao (2025), "The positivity technique and low-lying zeros of Dirichlet 
 L-functions", arXiv:2503.15832, Theorem 8, equation (14), page 6.
 [Barner's formulation of Weil's explicit formula]
"""

from e14c.filesystem import write_file
write_file("/workspace/downloads/explicit_formula_dirichlet_L.md", formula_doc)
print("Formula document saved.")
```

```python
# Let me also save the verification script
verification_script = '''#!/usr/bin/env python3
"""
Numerical verification of the Guinand-Weil explicit formula
for Dirichlet L-functions.

Formula verified:
 Sum_{rho} h((rho-1/2)/i) = g(0)*log(q/pi) 
 + (1/(2pi)) * int h(t) Re psi(1/4 + a_chi/2 + it/2) dt
 - 2 * Sum_{n>=2} Lambda(n) Re(chi(n)) n^{-1/2} g(log n)

where:
 g(u) = (1/(2pi)) int h(t) e^{-iut} dt
 a_chi = 0 if chi(-1)=+1 (even), a_chi = 1 if chi(-1)=-1 (odd)

Test function: h(t) = exp(-alpha*t^2)
 => g(u) = (1/(2*sqrt(pi*alpha))) * exp(-u^2/(4*alpha))
"""

import numpy as np
from scipy.special import digamma
from scipy.integrate import quad


def primes_up_to(N):
 sieve = [True] * (N+1)
 sieve[0] = sieve[1] = False
 for i in range(2, int(N**0.5)+1):
 if sieve[i]:
 for j in range(i*i, N+1, i):
 sieve[j] = False
 return [i for i in range(2, N+1) if sieve[i]]


def von_mangoldt_values(N):
 primes = primes_up_to(N)
 result = {}
 for p in primes:
 pk, logp = p, np.log(p)
 while pk <= N:
 result[pk] = logp
 pk *= p
 return result


def verify_explicit_formula(chi_func, q, a_chi, zeros, alpha=0.5, N_prime=10000):
 """Verify the Guinand-Weil explicit formula for L(s, chi).
 
 chi_func: function n -> chi(n) (can be complex)
 q: modulus
 a_chi: 0 for even, 1 for odd
 zeros: list of positive imaginary parts of non-trivial zeros
 alpha: parameter for test function h(t) = exp(-alpha*t^2)
 """
 h = lambda t: np.exp(-alpha * t**2)
 g = lambda u: (1.0/(2*np.sqrt(np.pi*alpha))) * np.exp(-u**2/(4*alpha))
 
 # Term 1: g(0) * log(q/pi)
 term1 = g(0) * np.log(q / np.pi)
 
 # Term 2: (1/(2pi)) * integral
 def integrand(t):
 z = 0.25 + a_chi/2 + 0.5j*t
 return h(t) * digamma(z).real
 
 term2, _ = quad(integrand, -100, 100, limit=200)
 term2 /= (2*np.pi)
 
 # Term 3: prime sum
 vm = von_mangoldt_values(N_prime)
 ps = 0.0
 for n, L in vm.items():
 chi_val = chi_func(n)
 if chi_val != 0:
 ps += L * np.real(chi_val) * n**(-0.5) * g(np.log(n))
 term3 = -2 * ps
 
 RHS = term1 + term2 + term3
 
 # LHS: sum over zeros (both +gamma and -gamma)
 LHS = sum(2 * h(gamma) for gamma in zeros)
 
 print(f" g(0)*log(q/pi) = {term1:.15f}")
 print(f" Digamma integral = {term2:.15f}")
 print(f" Prime sum = {term3:.15f}")
 print(f" RHS = {RHS:.15e}")
 print(f" LHS = {LHS:.15e}")
 print(f" |Difference| = {abs(LHS-RHS):.2e}")
 return RHS, LHS


if __name__ == "__main__":
 # 1. Verify for L(s, (./5)), EVEN character
 print("=" * 60)
 print("L(s, chi_5) -- Legendre symbol mod 5, EVEN")
 print("=" * 60)
 
 chi5 = lambda n: [0, 1, -1, -1, 1, 0][n % 5] if n % 5 != 0 else 0
 chi5_zeros = [
 6.64845334472771, 9.83144443288667, 11.9588456260835,
 16.0338211283842, 17.5669942923256, 19.5407326227848,
 ]
 verify_explicit_formula(chi5, 5, 0, chi5_zeros)
 
 # 2. Verify for L(s, chi_4), ODD character
 print()
 print("=" * 60)
 print("L(s, chi_4) -- odd character mod 4")
 print("=" * 60)
 
 chi4 = lambda n: [0, 1, 0, -1, 0][n % 4] if n % 4 in [1,3] else 0
 chi4_zeros = [6.02094890465140, 10.2437295954776, 12.5880432834498]
 verify_explicit_formula(chi4, 4, 1, chi4_zeros)
'''

write_file("/workspace/downloads/verify_explicit_formula.py", verification_script)
print("Verification script saved.")
```
