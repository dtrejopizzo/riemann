```python

# Check if any of these results contain actual computational data
with open("downloads/zenodo/20260327-010638-0390ff.jsonl", 'r') as f:
 zeta_data = json.load(f)

print(f"Found {len(zeta_data.get('hits', {}).get('hits', []))} zeta-related results\n")

# Look for datasets with actual data files (not just PDFs)
data_records = []
for hit in zeta_data.get('hits', {}).get('hits', []):
 title = hit.get('metadata', {}).get('title', '')
 files = hit.get('files', [])
 
 # Check if there are non-PDF files
 has_data = False
 for file in files:
 filename = file.get('key', '')
 if not filename.endswith('.pdf'):
 has_data = True
 break
 
 if has_data:
 data_records.append({
 'title': title,
 'doi': hit.get('doi', ''),
 'files': [f.get('key') for f in files]
 })

if data_records:
 print("Records with potential data files:")
 for rec in data_records:
 print(f"\n{rec['title']}")
 print(f"Files: {rec['files']}")
else:
 print("No records with non-PDF data files found.")
```
