```python

# Let me try a broader search on Zenodo for Dirichlet polynomials or zeta function
# datasets that might contain this type of analysis

import requests
import json

def search_zenodo(query_terms):
 """Search Zenodo with simplified query terms"""
 base_url = "https://zenodo.org/api/records"
 
 params = {
 'q': query_terms,
 'size': 20,
 'sort': 'bestmatch'
 }
 
 headers = {'Accept': 'application/json'}
 
 try:
 response = requests.get(base_url, params=params, headers=headers)
 response.raise_for_status()
 return response.json()
 except Exception as e:
 return {'error': str(e)}

# Try multiple search strategies
search_queries = [
 'Dirichlet polynomials analysis',
 'Liouville function numerical data',
 'zeta function Dirichlet analysis',
 'omega decomposition peaks',
 'inter-class energy ratio number theory'
]

print("Attempting broader Zenodo searches:")
print("="*80)

for query in search_queries:
 print(f"\nSearching: {query}")
 result = search_zenodo(query)
 
 if 'error' in result:
 print(f" ✗ Error: {result['error']}")
 elif 'hits' in result and result['hits']['total'] > 0:
 print(f" ✓ Found {result['hits']['total']} results")
 for hit in result['hits']['hits'][:3]: # Show first 3
 print(f" - {hit.get('metadata', {}).get('title', 'No title')}")
 print(f" DOI: {hit.get('doi', 'N/A')}")
 else:
 print(f" ✗ No results found")
```

```python

# The broad searches found many results but none appear to match the specific
# "report r2" data described in the query. Let me check the top zeta function
# results more carefully to see if any contain the type of data described

import requests

def get_zenodo_record_details(doi):
 """Get detailed information about a Zenodo record"""
 record_id = doi.split('zenodo.')[-1]
 url = f"https://zenodo.org/api/records/{record_id}"
 
 try:
 response = requests.get(url, headers={'Accept': 'application/json'})
 response.raise_for_status()
 return response.json()
 except Exception as e:
 return {'error': str(e)}

# Check the zeta function related records
zeta_dois = [
 '10.5281/zenodo.13747355', # Dirichlet Eta / Riemann Zeta
 '10.5281/zenodo.13369644', # Riemann Zeta and Dirichlet Eta
 '10.5281/zenodo.13626175' # Analog Theorem for Dirichlet Eta
]

print("Examining zeta function related Zenodo records in detail:")
print("="*80)

for doi in zeta_dois:
 print(f"\n{doi}")
 details = get_zenodo_record_details(doi)
 
 if 'error' in details:
 print(f" ✗ Error: {details['error']}")
 continue
 
 metadata = details.get('metadata', {})
 print(f" Title: {metadata.get('title', 'N/A')}")
 print(f" Description: {metadata.get('description', 'N/A')[:200]}...")
 
 # Check for files
 files = details.get('files', [])
 if files:
 print(f" Files ({len(files)}):")
 for f in files:
 print(f" - {f.get('key', 'N/A')} ({f.get('size', 0)} bytes)")
 
 # Check keywords
 keywords = metadata.get('keywords', [])
 if keywords:
 print(f" Keywords: {', '.join(keywords)}")
```
