```python
module_src = '''"""
Reconstructed weil_quadratic_form_general.py — provides _phi_at_points and _g_at_u
used by weil_modular_arch.compute_Q_modular.

Hermite-Gauss test-function convention (matches the explicit M_arch construction
visible in weil_modular_arch.py):

 psi_j(x) = H_j(x) * exp(-x^2/2) / sqrt( 2^j j! sqrt(pi) ) (L^2-orthonormal Hermite functions)
 phi_j(r) = (1/sqrt(sigma)) * psi_j( (r - T0)/sigma ) (centered at T0, width sigma)

Kernel:
 g_{ij}(u) = (1/(2*pi)) * integral_{-inf}^{inf} phi_i(r) phi_j(r) e^{i r u} dr
 = (e^{i T0 u}/(2*pi*sqrt(2^i i!) sqrt(2^j j!) sqrt(pi)))
 * integral H_i(x) H_j(x) exp(-x^2) e^{i sigma u x} dx
 evaluated by n_nodes_quad-point Gauss-Hermite quadrature.

Returned shapes match the calls inside weil_modular_arch.compute_Q_modular:
 _phi_at_points(points, sigma, T0, J) -> ndarray of shape (len(points), J)
 _g_at_u(u_vals, sigma, T0, J, n_nodes_quad) -> complex ndarray of shape (len(u_vals), J, J)
"""
import numpy as np
from scipy.special import gammaln, roots_hermite


def _phi_at_points(points, sigma, T0, J):
 points = np.asarray(points, dtype=np.float64)
 x = (points - T0) / sigma # (N,)
 N = x.size
 H = np.zeros((J, N))
 H[0] = 1.0
 if J > 1:
 H[1] = 2 * x
 for n in range(1, J - 1):
 H[n + 1] = 2 * x * H[n] - 2 * n * H[n - 1]
 # log(sqrt(2^j j! sqrt(pi))) = 0.5*(j*log2 + lgamma(j+1) + 0.5*log(pi))
 log_norm = 0.5 * (np.arange(J) * np.log(2.0) + gammaln(np.arange(J) + 1.0) + 0.5 * np.log(np.pi))
 # psi_j(x) = H_j(x) * exp(-x^2/2) / norm_j
 log_phase = -0.5 * x * x # (N,)
 psi = H * np.exp(log_phase[None, :] - log_norm[:, None]) # (J, N)
 # phi_j(r) = (1/sqrt(sigma)) * psi_j(x)
 phi = psi / np.sqrt(sigma)
 return phi.T # (N, J)


def _g_at_u(u_vals, sigma, T0, J, n_nodes_quad):
 u = np.asarray(u_vals, dtype=np.float64)
 Nu = u.size
 x, w = roots_hermite(n_nodes_quad) # ∫f(x)exp(-x^2)dx ≈ Σ w_k f(x_k)
 # Hermite polynomials at quadrature nodes
 H = np.zeros((J, n_nodes_quad))
 H[0] = 1.0
 if J > 1:
 H[1] = 2 * x
 for n in range(1, J - 1):
 H[n + 1] = 2 * x * H[n] - 2 * n * H[n - 1]
 # normalization: log( sqrt(2^j j!) ) — note: NO sqrt(pi) here because exp(-x^2) is
 # absorbed into the GH weight (we are working with H_j, not psi_j directly)
 log_norm_half = 0.5 * (np.arange(J) * np.log(2.0) + gammaln(np.arange(J) + 1.0))
 # exp_neg_norm[i,j] = 1/( sqrt(2^i i!) * sqrt(2^j j!) * sqrt(pi) )
 exp_neg_norm = np.exp(-(log_norm_half[:, None] + log_norm_half[None, :] + 0.5 * np.log(np.pi)))
 # Build complex exponential factors: e^{i sigma u x_k} -> shape (Nu, K)
 K = n_nodes_quad
 phase_inner = np.exp(1j * sigma * np.outer(u, x)) # (Nu, K)
 phase_outer = np.exp(1j * T0 * u) # (Nu,)
 # Weighted Hermite outer-product contraction over quadrature nodes:
 # I_{n,i,j}(u_n) = sum_k w_k * H_i(x_k) * H_j(x_k) * e^{i sigma u_n x_k}
 Hw = H * w[None, :] # (J, K)
 # For each u, compute (J,J) matrix M(u)_{ij} = sum_k (Hw_{i,k}) * phase_inner_{n,k} * H_{j,k}
 # Vectorize over n:
 # Build intermediate A_{n,i,k} = Hw_{i,k} * phase_inner_{n,k}
 A = Hw[None, :, :] * phase_inner[:, None, :] # (Nu, J, K)
 Mat = A @ H.T # (Nu, J, J)
 Mat = Mat * phase_outer[:, None, None] # multiply by e^{i T0 u}
 Mat = Mat * exp_neg_norm[None, :, :] / (2.0 * np.pi)
 return Mat
'''

with open('/workspace/54414482-5095-4d5b-a639-ad2f746dfde0/weil_quadratic_form_general.py','w') as f:
 f.write(module_src)
print("written", len(module_src), "bytes")
```

```python
# Run with arch_factor=1.0 and arch_factor=0.5 at the benchmark.
# Note: with only 61 zeros (up to ~103), at T0=85.7 sigma=2 J=10, the Hermite-Gauss
# basis is concentrated near |γ-T0| <= O(sigma*sqrt(J)) = O(2*sqrt(10)) ≈ 6.3 ;
# the relevant window is roughly 79 <= γ <= 92, all well within our zero list. Good.
import time

T0, sigma, J = 85.7, 2.0, 10
X = 10**4 # primes_cutoff

t0 = time.time()
Q1, parts1 = wma.compute_Q_modular(gammas, T0=T0, sigma=sigma, J=J,
 weight=12, conductor=1,
 primes_cutoff=X, n_nodes_quad=200,
 include_negative_zeros=True,
 arch_factor=1.0)
print("arch_factor=1.0 done in", round(time.time()-t0,1),"s")

t0 = time.time()
Q05, parts05 = wma.compute_Q_modular(gammas, T0=T0, sigma=sigma, J=J,
 weight=12, conductor=1,
 primes_cutoff=X, n_nodes_quad=200,
 include_negative_zeros=True,
 arch_factor=0.5)
print("arch_factor=0.5 done in", round(time.time()-t0,1),"s")

def tr(M): return float(np.trace(M))
tz = tr(parts1["M_zeros"])
ta1 = tr(parts1["M_arith"])
ta05 = tr(parts05["M_arith"])

print()
print("tr(M_zeros) =", tz)
print("tr(M_arith) [a=1.0] =", ta1)
print("tr(M_arith) [a=0.5] =", ta05)
print()
res1 = abs(tz - ta1) / abs(tz)
res05 = abs(tz - ta05) / abs(tz)
print(f"Relative trace residual at arch_factor=1.0: {res1:.6e} ({res1*100:.4f}%)")
print(f"Relative trace residual at arch_factor=0.5: {res05:.6e} ({res05*100:.4f}%)")
print()
print("tr(M_arch) =", tr(parts1["M_arch"]))
print("tr(M_arch)/2=", tr(parts05["M_arch"]))
print("tr(M_primes)=", tr(parts1["M_primes"]))
print("tr(M_polar) =", tr(parts1["M_polar"]))
```

```python
# Quantitative findings:
# - With arch_factor=1.0, residual ≈ 8.96%. This is even larger than the originally
# reported ~4.4% (which was with the older 12-zero-cutoff buggy list at X=10^4).
# Note our zero list still includes only 61 zeros (up to γ≈103); the missing high
# zeros add additional positive trace contributions which would *increase* tr(M_zeros)
# and reduce the residual. Therefore 8.96% is an over-estimate (upper bound) of the
# true residual; the true value once all zeros are added should drop.
# - With arch_factor=0.5, residual ≈ 83.9%. Catastrophically larger.
#
# Direction check: tr(M_arith) = M_arch - M_primes + M_polar
# = 14.17 - 0.0054 + (-5.85) = 8.31 at arch_factor=1.0
# tr(M_zeros) = 7.63 < 8.31, so reducing M_arch makes the gap WIDER (8.31 → 1.23, while
# target is 7.63). Halving M_arch over-corrects in the wrong way: it cuts off ~7.08 from
# the arithmetic side, pushing M_arith from ~104% of tr(M_zeros) down to ~16%.
#
# The hypothesis is rejected: scaling M_arch by 0.5 does NOT resolve the trace mismatch;
# it makes the relative residual ~9.4× larger.
#
# Caveat to be transparent about: my analysis used a reconstructed
# weil_quadratic_form_general.py (validated to ~1e-13 against direct quadrature for both
# φ orthonormality and the g_{ij}(u) kernel definition), and only 61 reliable zeros (up
# to γ≈103) rather than the full target list. With more zeros the unscaled residual
# should drop; the qualitative conclusion that arch_factor=0.5 is far worse than 1.0 is
# robust to additional zeros because M_zeros is monotone-increasing with more zeros and
# the arch_factor=0.5 trace is already short by ~6.4 (= 7.63 - 1.23), while adding
# trailing zeros at γ>>T0 contributes very little to tr(M_zeros) (phi^T phi decays
# Gaussian-fast outside the window).
#
# Verify: how much trace mass do the existing zeros contribute, and how much from those
# with γ outside the basis window [T0 - 3*σ*sqrt(J), T0 + 3*σ*sqrt(J)] ≈ [66.7,104.7]?
import numpy as np
window_low = T0 - 3*sigma*np.sqrt(J)
window_high = T0 + 3*sigma*np.sqrt(J)
print(f"Effective basis window: [{window_low:.2f}, {window_high:.2f}]")
in_window = (gammas >= window_low) & (gammas <= window_high)
out_window = ~in_window
# Compute M_zeros contribution from out-of-window zeros separately
import sys
m = sys.modules['weil_quadratic_form_general']
zarr = gammas.copy()
gout = zarr[out_window]
g_all = np.concatenate([gout, -gout])
phi = m._phi_at_points(g_all, sigma, T0, J)
M_out = phi.T @ phi
print(f"trace contribution to M_zeros from out-of-window (|γ|>{window_high:.1f}): {np.trace(M_out):.3e}")
print(f"as fraction of full tr(M_zeros): {np.trace(M_out)/np.trace(parts1['M_zeros']):.3e}")
```

```python
# Excellent. Out-of-window zeros contribute negligibly (≲1e-27 fractional). The 61 zeros
# we have already span the relevant window (66.7..104.7) since our list goes to γ≈103.
# Let's check if any zeros inside (104.7, ~110) are missing — they could be relevant.
# Our list ends at 103.17, just below the upper bound 104.67. Let's check how many zeros
# we actually have inside the window:
print(f"Zeros in window: {in_window.sum()} of {len(gammas)}")
print("γ values just below upper window edge:", gammas[gammas > T0+2*sigma*np.sqrt(J)])
# Two more zeros immediately after 103.17 (~104, ~105) would lie in the window.
# Bound their effect: the next zero after 103.17 in LMFDB for L(Δ,s) lies around ~107.
# Even ONE missed zero just inside the upper window would contribute O(phi^T phi at γ=104) which is small.
# Quantify: synthetic test - add a hypothetical zero at γ=104.5 (just inside window upper edge):
hypo = np.append(gammas, 104.5)
g_all2 = np.concatenate([hypo, -hypo])
m = sys.modules['weil_quadratic_form_general']
phi = m._phi_at_points(g_all2, sigma, T0, J)
M_hypo = phi.T @ phi
print(f"tr(M_zeros) with hypothetical extra zero at γ=104.5: {np.trace(M_hypo):.6f}")
print(f"vs. current tr(M_zeros): {np.trace(parts1['M_zeros']):.6f}")
print(f"Δ: {np.trace(M_hypo) - np.trace(parts1['M_zeros']):.3e}")
```

```python
# A zero at γ=104.5 adds essentially nothing to tr(M_zeros). So the 8.96% residual is
# numerically *robust* to extending the zero list — the basis window only "sees" zeros
# up to about γ≈101. The residual is not caused by missing zeros.
#
# Therefore the hypothesis "arch_factor=0.5 resolves the trace mismatch" is definitively
# refuted by this calculation. The mismatch is real (~9% with the bona fide explicit-
# formula assembly) and halving M_arch makes it 9× worse.
#
# Let me also test a few alternate scalings just to characterize: factor that minimizes residual?
from scipy.optimize import minimize_scalar
M_arch_full = parts1["M_arch"]
M_primes = parts1["M_primes"]
M_polar = parts1["M_polar"]
M_zeros = parts1["M_zeros"]
def residual(a):
 M_arith = M_polar + a*M_arch_full - M_primes
 return abs(np.trace(M_zeros) - np.trace(M_arith)) / abs(np.trace(M_zeros))
xs = np.linspace(0.0, 1.5, 31)
ys = [residual(a) for a in xs]
for a,r in zip(xs,ys):
 print(f"arch_factor={a:.2f} |resid|={r:.4e}")
opt = minimize_scalar(residual, bounds=(0,2), method='bounded')
print(f"\noptimum arch_factor that zeros the trace residual: {opt.x:.4f}, resid={opt.fun:.3e}")
```
