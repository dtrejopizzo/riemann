```python
# 0.10s per evaluation at dps=30. Way faster than dps=50.
# Let me write a fast custom finder
def find_zeros_fast(N_zeros, t_start=8.0, step=0.3, y_max=10, dps=30):
 mp.mp.dps = dps
 zeros = []
 t_prev = mp.mpf(t_start)
 Lp = Lambda_fast(t_prev, y_max)
 t = t_prev + step
 while len(zeros) < N_zeros:
 Lc = Lambda_fast(t, y_max)
 if Lp * Lc < 0:
 a_, b_ = t_prev, t
 La = Lp
 for _ in range(80):
 m_ = (a_ + b_) / 2
 Lm = Lambda_fast(m_, y_max)
 if Lm == 0 or abs(b_-a_) < mp.mpf(10)**(-dps+10):
 break
 if La * Lm < 0:
 b_ = m_
 else:
 a_, La = m_, Lm
 zeros.append((a_+b_)/2)
 t_prev, Lp = t, Lc
 t += step
 return zeros

t0 = time.time()
zfast = find_zeros_fast(10)
print(f"10 zeros in {time.time()-t0:.1f}s")
for z, z_old in zip(zfast, zeros20[:10]):
 print(float(z), float(z_old), float(z)-float(z_old))
```

```python
# Great. τ(p) values correct. Now build engine for L(Δ).
# I'll add a new family "modular" with parameter k=12.
# Rather than modifying the file, I'll write a derived class/function in the notebook.

import numpy as np
from scipy.special import digamma, gammaln, roots_hermite
from sympy import primerange

def _phi_at_points(t_points, sigma, T0, J):
 t = np.asarray(t_points, dtype=np.float64)
 u = (t - T0) / sigma
 N = len(t)
 h = np.zeros((J, N))
 h[0] = np.pi ** (-0.25) * np.exp(-u * u / 2)
 if J > 1:
 h[1] = np.sqrt(2.0) * u * h[0]
 for n in range(1, J - 1):
 h[n + 1] = np.sqrt(2.0 / (n + 1)) * u * h[n] - np.sqrt(n / (n + 1)) * h[n - 1]
 return (h / np.sqrt(sigma)).T

def _g_at_u(u_vals, sigma, T0, J, n_nodes_quad):
 """Compute g_{ij}(u) = ∫ φ_i(t) φ_j(t) e^{iut} dt for each u — but for our setup we need
 g_B(u) used to compute prime-sum contributions. The implementation returns a complex 
 matrix per u (the (i,j) element)."""
 # For real-valued localized basis: φ_i is real, so g_ij(u) = ∫ φ_i(t)φ_j(t) e^{iut} dt
 # Since φ_i(t) = (1/√σ) h_i((t-T0)/σ) where h_i is Hermite, 
 # g_ij(u) = e^{iu T0} ∫ (1/σ) h_i(τ)h_j(τ) e^{iuστ} σ dτ (τ=(t-T0)/σ)
 # = e^{iu T0} ∫ h_i(τ) h_j(τ) e^{iuστ} dτ
 # Use Gauss-Hermite: ∫ p(τ) e^{-τ²} dτ but our h_i are normalized so h_i(τ) = e^{-τ²/2} H_i(τ)/√(2^i i!√π)
 # So h_i(τ)h_j(τ) = e^{-τ²} H_i H_j / norm. Then the integral becomes a Hermite quadrature.
 x, w = roots_hermite(n_nodes_quad)
 H = np.zeros((J, n_nodes_quad))
 H[0] = 1.0
 if J > 1:
 H[1] = 2 * x
 for n in range(1, J - 1):
 H[n+1] = 2 * x * H[n] - 2 * n * H[n-1]
 norm = np.array([0.5 * (k * np.log(2) + gammaln(k + 1)) for k in range(J)])
 norm_ij = norm[:, None] + norm[None, :] + 0.5 * np.log(np.pi)
 exp_neg_norm = np.exp(-norm_ij)
 # g_ij(u) = e^{iu T0} (1/√π) ... wait let me re-derive normalization
 # h_n(τ) = (2^n n! √π)^{-1/2} H_n(τ) e^{-τ²/2}
 # h_i h_j = (2^{i+j} i! j! π)^{-1/2} H_i H_j e^{-τ²}
 # ∫ h_i h_j e^{iuστ} dτ = (...)^{-1/2} ∫ H_i H_j e^{-τ²} e^{iuστ} dτ
 # Approximate via Gauss-Hermite (quadrature on e^{-τ²} weight):
 # ≈ Σ_q w_q H_i(x_q) H_j(x_q) e^{iuσx_q}
 # g_ij(u) = e^{iu T0} · exp(-norm_ij) · Σ_q w_q H_i(x_q) H_j(x_q) e^{iuσx_q}
 n_u = len(u_vals)
 g = np.zeros((n_u, J, J), dtype=np.complex128)
 for idx, u in enumerate(u_vals):
 ph = np.exp(1j * u * sigma * x) * w # (n_nodes_quad,)
 HP = H * ph[None, :]
 mat = HP @ H.T
 g[idx] = np.exp(1j * u * T0) * exp_neg_norm * mat
 return g

print("Helpers defined")
```

```python
# Let me add a 'modular' family by extending the existing engine. Write a new file.

modular_code = '''"""
Extension of weil_quadratic_form_general.py to support modular L-functions L(f, s)
for weight-k level-1 holomorphic cusp forms (e.g. Ramanujan Δ, k=12).

Analytic normalization: critical line Re(s)=1/2; Dirichlet coefficients
 a_f(n) = c_f(n) / n^{(k-1)/2}, with c_f the unnormalized arithmetic coefficients.
For Δ: c_Δ(n) = τ(n), so a_Δ(n) = τ(n)/n^{11/2}, |a_Δ(p)| ≤ 2 (Deligne).

Explicit formula for L(Δ, s) (degree 2, conductor 1, no pole, Γ_C(s + 11/2)):
 Σ_ρ h(γ) = (1/2π) ∫ h(r) [-2 log(2π) + 2 Re ψ(6 + ir)] dr
 - 2 Σ_{p^k} (α_p^k + β_p^k) log(p) / p^{k/2} · g(k log p)
where α_p + β_p = a_Δ(p), α_p β_p = 1 (Deligne), giving Recurrence
 c_k(p) = α^k + β^k: c_0=2, c_1 = a_Δ(p), c_k = a_Δ(p) c_{k-1} - c_{k-2}.

Implementation mirrors weil_quadratic_form_general.compute_Q:
 M_arith = M_polar + M_arch - M_primes; Q = M_zeros - M_arith.
"""
import numpy as np
from scipy.special import digamma, gammaln, roots_hermite
from sympy import primerange
from weil_quadratic_form_general import _phi_at_points, _g_at_u


def _tau_coeffs(N):
 """Compute τ(1),...,τ(N) via Δ = q ∏(1-q^n)^24."""
 from math import comb
 c = [0] * (N + 1)
 c[0] = 1
 for n in range(1, N + 1):
 new_c = [0] * (N + 1)
 for i, ci in enumerate(c):
 if ci == 0:
 continue
 for k in range(25):
 idx = i + n * k
 if idx > N:
 break
 new_c[idx] += ci * comb(24, k) * (-1) ** k
 c = new_c
 return [c[n - 1] for n in range(1, N + 1)]


def compute_Q_modular(zeros, T0, sigma, J, weight=12, conductor=1,
 primes_cutoff=100000, n_nodes_quad=200,
 include_negative_zeros=True, tau_func=None):
 """Compute Q = M_zeros - M_arith for a weight-k level-N holomorphic
 cusp form L-function (e.g. L(Δ, s) for k=12, N=1).

 Parameters
 ----------
 zeros : list/array of real γ (analytic normalization, ρ = 1/2 + iγ)
 T0, sigma, J : Hermite-window parameters
 weight : holomorphic weight (Δ uses 12)
 conductor : level N (Δ uses 1)
 primes_cutoff : X for prime-power sum
 n_nodes_quad : Gauss-Hermite nodes
 include_negative_zeros : pair γ with -γ (self-dual real form, true for Δ)
 tau_func : optional callable n→τ(n); if None, computed from Δ q-expansion up to primes_cutoff
 """
 k = weight
 half_kminus = (k - 1) / 2.0 # =11/2 for Δ

 # ---- M_zeros ----
 zeros_arr = np.asarray(zeros, dtype=np.float64)
 if include_negative_zeros:
 g_all = np.concatenate([zeros_arr, -zeros_arr])
 else:
 g_all = zeros_arr
 phi = _phi_at_points(g_all, sigma, T0, J)
 M_zeros = phi.T @ phi

 # ---- Build prime coefficient list ----
 primes_list = list(primerange(2, primes_cutoff + 1))
 # Get τ(p) for each prime; need τ(p^k) handled via recurrence using a_Δ(p) = τ(p)/p^{11/2}
 if tau_func is None:
 # Compute τ for primes up to primes_cutoff
 max_p = primes_list[-1] if primes_list else 2
 tau_vals = _tau_coeffs(max_p)
 def tau_func(n):
 return tau_vals[n - 1]

 u_vals = []
 weights = []
 for p in primes_list:
 a_p = tau_func(p) / (p ** half_kminus) # analytic-norm Dirichlet coeff
 # Recurrence for c_k(p) = α_p^k + β_p^k
 c_prev2 = 2.0 # c_0
 c_prev1 = a_p # c_1
 pk = p
 kk = 1
 # First term: k=1
 if pk <= primes_cutoff:
 u_vals.append(np.log(p))
 weights.append(c_prev1 * np.log(p) / np.sqrt(pk))
 kk = 2
 pk = p * p
 while pk <= primes_cutoff:
 c_cur = a_p * c_prev1 - c_prev2
 u_vals.append(kk * np.log(p))
 weights.append(c_cur * np.log(p) / np.sqrt(pk))
 c_prev2 = c_prev1
 c_prev1 = c_cur
 kk += 1
 pk *= p
 u_vals = np.array(u_vals)
 weights = np.array(weights, dtype=np.float64)

 # ---- M_primes ----
 # Existing Dirichlet convention: M_primes_ij = Re[Σ w_n g_ij(log n)] / π
 # For modular: same convention, single-side prime sum corresponds to
 # the standard "-2 Σ Λ_L(n)/√n g(log n)" via the symmetry of the basis.
 # Since c_k(p) is real (Δ is self-dual), no Re needed but harmless.
 M_primes = np.zeros((J, J))
 chunk = 2000
 for i in range(0, len(u_vals), chunk):
 u_chunk = u_vals[i:i + chunk]
 w_chunk = weights[i:i + chunk]
 g_chunk = _g_at_u(u_chunk, sigma, T0, J, n_nodes_quad)
 M_primes += np.real(np.sum(w_chunk[:, None, None] * g_chunk, axis=0)) / np.pi

 # ---- M_arch ----
 # Gamma factor Γ_C(s + 11/2): log-derivative = -log(2π) + ψ(s + (k-1)/2)
 # Symmetric integrand: (-log(2π) + Re ψ(1/2+ir + (k-1)/2)) doubled (since Λ(s)=Λ(1-s)),
 # but for degree-2 form we have ONE Γ_C, which is "two" Γ_R's, giving the factor 2.
 # In the explicit formula form Σ_ρ h(γ) = (1/2π) ∫ h(r) Φ(r) dr - 2 Σ ...
 # Φ(r) = log N + (γ_∞)'/γ_∞(1/2+ir) + (γ_∞)'/γ_∞(1/2-ir)
 # For γ_∞(s)=Γ_C(s+(k-1)/2):
 # Φ(r) = log N + 2 Re[-log(2π) + ψ(1/2+ir+(k-1)/2)]
 # = log N - 2 log(2π) + 2 Re ψ(k/2 + ir)
 x, w = roots_hermite(n_nodes_quad)
 H = np.zeros((J, n_nodes_quad))
 H[0] = 1.0
 if J > 1:
 H[1] = 2 * x
 for n in range(1, J - 1):
 H[n+1] = 2 * x * H[n] - 2 * n * H[n-1]
 norm = np.array([0.5 * (kk_ * np.log(2) + gammaln(kk_ + 1)) for kk_ in range(J)])
 norm_ij = norm[:, None] + norm[None, :] + 0.5 * np.log(np.pi)
 exp_neg_norm = np.exp(-norm_ij)
 t_nodes = T0 + sigma * x
 # ψ(k/2 + i t); take real part, multiply by 2:
 arg = (k / 2.0) + 1j * t_nodes
 psi_vals = 2.0 * np.real(digamma(arg))
 diag = w * psi_vals
 HP = H * diag[None, :]
 M_arch = (HP @ H.T) * exp_neg_norm / (2 * np.pi)

 # ---- M_polar ----
 # log N - 2 log(2π); for Δ N=1 so log(1) = 0
 polar_const = np.log(conductor) - 2.0 * np.log(2.0 * np.pi)
 M_polar = (polar_const / (2.0 * np.pi)) * np.eye(J)

 M_arith = M_polar + M_arch - M_primes
 Q = M_zeros - M_arith
 Q = 0.5 * (Q + Q.T)
 return Q, {
 "M_zeros": M_zeros,
 "M_primes": M_primes,
 "M_arch": M_arch,
 "M_polar": M_polar,
 "M_arith": M_arith,
 }
'''

with open('weil_modular.py', 'w') as f:
 f.write(modular_code)
print("Wrote weil_modular.py")
```
