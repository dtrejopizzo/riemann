```python

# 3. Demonstrate Davenport-Heilbronn coefficients
print("\n3. DAVENPORT-HEILBRONN FUNCTION COEFFICIENTS")

# Define kappa
sqrt5 = np.sqrt(5)
kappa = (sqrt5 - 1) / (2 * np.sqrt(5 * (sqrt5 - 1)))
print(f" κ = {kappa:.10f}")
print(f" θ = arctan(κ) = {np.arctan(kappa):.10f} radians")

# Compute first few DH coefficients
def chi_order4(n):
 """Complex character mod 5 of order 4"""
 residue = n % 5
 return {0: 0, 1: 1, 2: 1j, 3: -1j, 4: -1}[residue]

def dh_coefficient(n, kappa):
 """n-th Dirichlet coefficient of Davenport-Heilbronn function"""
 chi_n = chi_order4(n)
 chi_bar_n = np.conj(chi_n)
 a_n = ((1 - 1j*kappa) / 2) * chi_n + ((1 + 1j*kappa) / 2) * chi_bar_n
 return a_n

print(f"\n First 20 DH coefficients:")
for n in range(1, 21):
 a_n = dh_coefficient(n, kappa)
 print(f" a_{n:2d} = {a_n.real:8.5f} + {a_n.imag:8.5f}i")

print(f"\n Properties:")
print(f" - Coefficients are COMPLEX")
print(f" - NOT multiplicative: a_mn ≠ a_m · a_n in general")
print(f" - Storage: complex128 array of length N")
```

```python

# 5. Demonstrate the central computational object: D_F(t; N)
print("\n5. CENTRAL COMPUTATIONAL OBJECT: D_F(t; N)")
print(" The Resonance Detector")

print(f"\n Definition:")
print(f" D_F(t; N) = Σ_(n≤N) a_n(F) / n^(1/2 + it)")
print(f" = Σ_(n≤N) a_n(F) · n^(-1/2) · exp(-it·log(n))")

print(f"\n Computational challenges:")
print(f" 1. Highly oscillatory sum with ~N terms")
print(f" 2. Massive cancellation → catastrophic round-off error if naive")
print(f" 3. Must use compensated summation (Kahan or pairwise)")
print(f" 4. For validation: compare float64 vs mpmath arbitrary precision")

# Demonstrate simple computation for small N
def compute_D_F_naive(a_coeffs, t, N):
 """Compute D_F(t; N) - NAIVE (for demonstration only)"""
 result = 0.0 + 0.0j
 for n in range(1, N+1):
 term = a_coeffs[n] * n**(-0.5) * np.exp(-1j * t * np.log(n))
 result += term
 return result

# For ζ: a_n = 1 for all n
N_small = 1000
t_test = 10.0
a_zeta = {n: 1.0 for n in range(1, N_small+1)}
D_zeta = compute_D_F_naive(a_zeta, t_test, N_small)

print(f"\n Example: D_ζ(t={t_test}, N={N_small})")
print(f" Real part: {D_zeta.real:.6f}")
print(f" Imag part: {D_zeta.imag:.6f}")
print(f" Magnitude: {np.abs(D_zeta):.6f}")
print(f" Phase: {np.angle(D_zeta):.6f} radians")

print(f"\n Data structure:")
print(f" - Input: a_n coefficients (array of length N)")
print(f" - Input: t values (grid of heights, typically 10^3 to 10^5 points)")
print(f" - Output: D_F(t; N) for each t (complex128 array)")
print(f" - Typical size: N=10^6, num_t=10^4 → 10^10 arithmetic operations")
```

```python

# 6. Demonstrate Kahan compensated summation (REQUIRED for numerical stability)
print("\n6. NUMERICAL PRECISION: KAHAN COMPENSATED SUMMATION")

def kahan_sum_complex(values):
 """Kahan compensated summation for complex array"""
 s = 0.0 + 0.0j
 c = 0.0 + 0.0j # Running compensation
 
 for val in values:
 y = val - c
 t = s + y
 c = (t - s) - y
 s = t
 return s

def compute_D_F_kahan(a_coeffs, t, N):
 """Compute D_F(t; N) with Kahan summation - RECOMMENDED"""
 terms = []
 for n in range(1, N+1):
 term = a_coeffs[n] * n**(-0.5) * np.exp(-1j * t * np.log(n))
 terms.append(term)
 return kahan_sum_complex(terms)

# Compare naive vs Kahan for larger sum
N_medium = 5000
t_test = 100.0
a_zeta_medium = {n: 1.0 for n in range(1, N_medium+1)}

D_naive = compute_D_F_naive(a_zeta_medium, t_test, N_medium)
D_kahan = compute_D_F_kahan(a_zeta_medium, t_test, N_medium)

print(f"\n Test case: D_ζ(t={t_test}, N={N_medium})")
print(f" Naive summation: {D_naive.real:.10f} + {D_naive.imag:.10f}i")
print(f" Kahan summation: {D_kahan.real:.10f} + {D_kahan.imag:.10f}i")
print(f" Difference: {abs(D_kahan - D_naive):.2e}")

print(f"\n For production (N=10^6, T=10^4):")
print(f" - MUST use Kahan or pairwise summation")
print(f" - For N ≤ 10^4: validate against mpmath (arbitrary precision)")
print(f" - Required agreement: ≥ 12 significant digits")
print(f" - Documentation: record summation method in metadata")
```
