```python

# Step 1: Import necessary libraries and define coefficient functions
import numpy as np
import mpmath
from scipy import signal
from scipy.stats import rayleigh
import matplotlib.pyplot as plt
from collections import defaultdict
import time

# Set random seed for reproducibility
np.random.seed(42)

print("Step 1: Defining coefficient functions for all function classes")
print("=" * 80)

# Helper: Generate primes up to N using Sieve of Eratosthenes
def sieve_of_eratosthenes(limit):
 """Generate all primes up to limit."""
 is_prime = np.ones(limit + 1, dtype=bool)
 is_prime[0:2] = False
 for i in range(2, int(limit**0.5) + 1):
 if is_prime[i]:
 is_prime[i*i::i] = False
 return np.where(is_prime)[0]

# Generate primes up to 10^6
N_max = 10**6
primes = sieve_of_eratosthenes(N_max)
print(f"Generated {len(primes)} primes up to {N_max:,}")

# Define χ₄: non-principal real character mod 5
# χ₄(1) = 1, χ₄(2) = -1, χ₄(3) = -1, χ₄(4) = 1, χ₄(0) = 0
def chi_4(n):
 """Non-principal real character mod 5."""
 if n % 5 == 0:
 return 0
 elif n % 5 == 1:
 return 1
 elif n % 5 == 2:
 return -1
 elif n % 5 == 3:
 return -1
 elif n % 5 == 4:
 return 1

# Define χ: primitive complex character mod 5 of order 4
# χ(1) = 1, χ(2) = i, χ(3) = -i, χ(4) = -1, χ(0) = 0
def chi_complex(n):
 """Primitive complex character mod 5 of order 4."""
 if n % 5 == 0:
 return 0
 elif n % 5 == 1:
 return 1
 elif n % 5 == 2:
 return 1j
 elif n % 5 == 3:
 return -1j
 elif n % 5 == 4:
 return -1

# Davenport-Heilbronn coefficients
# κ = (√5 - 1) / (2√(5(√5 - 1)))
sqrt5 = np.sqrt(5)
kappa = (sqrt5 - 1) / (2 * np.sqrt(5 * (sqrt5 - 1)))
print(f"Davenport-Heilbronn parameter κ = {kappa:.10f}")

# L_DH(s) = ((1-iκ)/2) L(s,χ) + ((1+iκ)/2) L(s,χ̄)
# a_n^{DH} = ((1-iκ)/2) χ(n) + ((1+iκ)/2) χ̄(n)
def a_DH(n):
 """Davenport-Heilbronn coefficient."""
 chi_n = chi_complex(n)
 chi_bar_n = np.conj(chi_n)
 return ((1 - 1j*kappa)/2) * chi_n + ((1 + 1j*kappa)/2) * chi_bar_n

# Test coefficients
print("\nCoefficient examples:")
for n in [1, 2, 3, 4, 5, 6, 7, 8]:
 a_zeta = 1
 a_chi4 = chi_4(n)
 a_dh = a_DH(n)
 print(f" n={n}: ζ={a_zeta}, χ₄={a_chi4:+d}, DH={a_dh}")

print("\nVerifying Davenport-Heilbronn coefficients are NOT multiplicative:")
# Check a_6 vs a_2 * a_3
a_6 = a_DH(6)
a_2 = a_DH(2)
a_3 = a_DH(3)
print(f" a_DH(6) = {a_6}")
print(f" a_DH(2) * a_DH(3) = {a_2 * a_3}")
print(f" Multiplicative? {np.abs(a_6 - a_2*a_3) < 1e-10}")
```

```python

# Step 2: Generate random multiplicative function coefficients
print("Step 2: Generating random multiplicative function")
print("=" * 80)

# For random multiplicative function: a_p = ±1 i.i.d. at primes, extended multiplicatively
# We need to store prime values and compute composite values on-the-fly

# Generate random signs for all primes up to N_max
prime_signs = {}
np.random.seed(42) # For reproducibility
for p in primes:
 prime_signs[p] = np.random.choice([-1, 1])

print(f"Generated random signs for {len(prime_signs)} primes")
print(f"First 10 prime signs: {[(p, prime_signs[p]) for p in primes[:10]]}")

# Function to compute a_rand(n) for random multiplicative function
def a_rand(n):
 """Random multiplicative function coefficient."""
 if n == 1:
 return 1
 
 # Factor n and multiply prime values
 result = 1
 temp_n = n
 for p in primes:
 if p * p > temp_n:
 if temp_n > 1: # temp_n is prime
 result *= prime_signs.get(temp_n, 1)
 break
 while temp_n % p == 0:
 result *= prime_signs[p]
 temp_n //= p
 return result

# Pre-compute all coefficients up to N_max for efficiency
print("\nPre-computing all coefficients up to N_max...")
start = time.time()

# Pre-compute coefficient arrays
coeff_zeta = np.ones(N_max + 1, dtype=float)
coeff_chi4 = np.array([chi_4(n) for n in range(N_max + 1)], dtype=float)
coeff_DH = np.array([a_DH(n) for n in range(N_max + 1)], dtype=complex)
coeff_rand = np.array([a_rand(n) for n in range(N_max + 1)], dtype=float)

elapsed = time.time() - start
print(f"Pre-computed all coefficients in {elapsed:.2f} seconds")

# Verify random multiplicativity
print("\nVerifying random multiplicative property:")
for (a, b) in [(2, 3), (5, 7), (11, 13)]:
 n = a * b
 a_n = coeff_rand[n]
 a_a = coeff_rand[a]
 a_b = coeff_rand[b]
 print(f" a_rand({a}*{b}) = {a_n}, a_rand({a})*a_rand({b}) = {a_a*a_b}, match: {a_n == a_a*a_b}")
```

```python

# Step 3: Implement D_F(t; N) computation with Kahan compensated summation
print("Step 3: Implementing D_F(t; N) with Kahan compensated summation")
print("=" * 80)

def kahan_sum_complex(values):
 """
 Kahan compensated summation for complex values.
 Reduces cumulative floating-point error from O(ε*N) to O(ε*log(N)).
 """
 s = 0.0 + 0.0j
 c = 0.0 + 0.0j # Running compensation for lost low-order bits
 
 for val in values:
 y = val - c
 t = s + y
 c = (t - s) - y
 s = t
 
 return s

def compute_D_F(t, N, coeffs):
 """
 Compute D_F(t; N) = Σ_{n≤N} a_n(F) / n^(1/2 + it)
 = Σ_{n≤N} a_n(F) * n^(-1/2) * exp(-it*log(n))
 
 Uses Kahan compensated summation for numerical stability.
 
 Parameters:
 - t: height parameter (real)
 - N: truncation length (integer)
 - coeffs: array of coefficients a_n for n=0,1,...,N
 
 Returns:
 - Complex value D_F(t; N)
 """
 n_values = np.arange(1, N + 1)
 
 # Compute n^(-1/2) * exp(-it*log(n))
 # = exp(-0.5*log(n) - it*log(n))
 # = exp(-(0.5 + it)*log(n))
 log_n = np.log(n_values)
 factors = np.exp(-(0.5 + 1j*t) * log_n)
 
 # Multiply by coefficients
 terms = coeffs[1:N+1] * factors
 
 # Use Kahan summation
 result = kahan_sum_complex(terms)
 
 return result

# Test on small example
print("Testing D_F computation on small example (N=10, t=1.0):")
for name, coeffs in [('ζ', coeff_zeta), ('χ₄', coeff_chi4), ('DH', coeff_DH)]:
 D = compute_D_F(1.0, 10, coeffs)
 print(f" D_{name}(1.0; 10) = {D}")
```

```python

# Step 4: Validate numerical accuracy against mpmath (mandatory!)
print("Step 4: MANDATORY VALIDATION - Comparing float64 vs mpmath")
print("=" * 80)

# Use mpmath for arbitrary precision validation
mpmath.mp.dps = 30 # 30 decimal places

def compute_D_F_mpmath(t, N, coeffs):
 """
 Compute D_F(t; N) using mpmath arbitrary precision.
 Used for validation only.
 """
 result = mpmath.mpc(0, 0)
 
 for n in range(1, N + 1):
 a_n = coeffs[n]
 # Convert to mpmath
 if isinstance(a_n, complex):
 a_n_mp = mpmath.mpc(a_n.real, a_n.imag)
 else:
 a_n_mp = mpmath.mpf(a_n)
 
 # n^(-1/2 - it) = exp(-(1/2 + it) * log(n))
 exponent = -(mpmath.mpf(0.5) + mpmath.mpc(0, t)) * mpmath.log(n)
 term = a_n_mp * mpmath.exp(exponent)
 result += term
 
 return result

# Validate for N = 10^4 (as required) at several t values
N_validate = 10**4
t_values_validate = [10.0, 50.0, 100.0]

print(f"Validating at N = {N_validate:,} for t values: {t_values_validate}")
print()

validation_results = []

for t_val in t_values_validate:
 print(f"t = {t_val}:")
 
 for name, coeffs in [('ζ', coeff_zeta), ('χ₄', coeff_chi4), ('DH', coeff_DH)]:
 # Compute with float64 + Kahan
 D_float = compute_D_F(t_val, N_validate, coeffs)
 
 # Compute with mpmath
 D_mp = compute_D_F_mpmath(t_val, N_validate, coeffs)
 
 # Compute relative error
 abs_D_mp = abs(D_mp)
 abs_error = abs(complex(D_mp) - D_float)
 rel_error = abs_error / abs_D_mp if abs_D_mp > 0 else 0
 
 validation_results.append({
 't': t_val,
 'function': name,
 'rel_error': float(rel_error)
 })
 
 # Check if relative error < 10^-12 (requirement)
 status = "PASS" if rel_error < 1e-12 else "FAIL"
 
 print(f" {name:3s}: |D| = {abs_D_mp:.6e}, rel_error = {rel_error:.6e} [{status}]")
 
 print()

# Check if all validations passed
all_passed = all(r['rel_error'] < 1e-12 for r in validation_results)

print("=" * 80)
print(f"VALIDATION RESULT: {'ALL PASSED ✓' if all_passed else 'FAILED ✗'}")
print("=" * 80)

if not all_passed:
 print("WARNING: Numerical validation failed. Data generation cannot proceed.")
 print("This would invalidate the entire dataset per research program requirements.")
else:
 print("Numerical accuracy validated. Proceeding with full dataset generation.")
```

```python

# Step 7: Extract prime phases at peak locations and perform Rayleigh test
print("\nStep 7: Extracting prime phases and performing Rayleigh uniformity test")
print("=" * 80)

# For each peak location, we need to compute:
# θ_p(t) = -t·log(p) + arg(a_p) for all primes p ≤ N

def extract_prime_phases(t, primes, coeff_func, N_max):
 """
 Extract phases θ_p(t) = -t·log(p) + arg(a_p) for all primes p ≤ N_max.
 
 Parameters:
 - t: height parameter
 - primes: array of prime numbers
 - coeff_func: function that returns coefficient a_p for prime p
 - N_max: maximum prime to consider
 
 Returns:
 - phases: array of phases in [0, 2π)
 """
 # Filter primes ≤ N_max
 primes_filtered = primes[primes <= N_max]
 
 phases = np.zeros(len(primes_filtered))
 
 for i, p in enumerate(primes_filtered):
 a_p = coeff_func(p)
 
 # Compute phase: -t·log(p) + arg(a_p)
 # For real a_p: arg(a_p) = 0 if a_p > 0, π if a_p < 0
 # For complex a_p: use np.angle
 if isinstance(a_p, complex):
 arg_a_p = np.angle(a_p)
 else:
 arg_a_p = 0 if a_p >= 0 else np.pi
 
 theta_p = -t * np.log(p) + arg_a_p
 
 # Reduce to [0, 2π)
 theta_p = theta_p % (2 * np.pi)
 
 phases[i] = theta_p
 
 return phases

# Rayleigh test for uniformity
# Test statistic: R = |Σ exp(iθ)| / n
# Under uniformity, 2nR² ~ χ²(2) for large n
# We use scipy's implementation which returns p-value

from scipy.stats import circmean, circstd

def rayleigh_test(phases):
 """
 Rayleigh test for uniformity of circular data.
 
 Returns:
 - R: mean resultant length (0 = uniform, 1 = perfectly aligned)
 - p_value: probability that uniform distribution would produce R this large
 """
 n = len(phases)
 
 # Compute mean resultant vector
 C = np.mean(np.cos(phases))
 S = np.mean(np.sin(phases))
 
 # Mean resultant length
 R = np.sqrt(C**2 + S**2)
 
 # Test statistic for uniformity
 # Under H0 (uniform), 2nR² ~ χ²(2) for large n
 test_stat = 2 * n * R**2
 
 # P-value from chi-squared distribution with 2 df
 from scipy.stats import chi2
 p_value = 1 - chi2.cdf(test_stat, df=2)
 
 return R, p_value

print("Analyzing phase distributions at peak locations...")
print()

# Test coefficient functions for each function class
coeff_functions = {
 'zeta': lambda p: 1.0,
 'chi4': chi_4,
 'DH': a_DH,
 'rand': lambda p: prime_signs.get(p, 1)
}

# Store results for all peaks
rayleigh_results = {}

for func_name in ['zeta', 'chi4', 'DH', 'rand']:
 print(f"\n{func_name.upper()}:")
 print("-" * 80)
 
 peak_data = results_peaks[func_name]
 t_peaks = peak_data['t_values']
 
 rayleigh_results[func_name] = []
 
 # Analyze top 5 peaks in detail (due to computational cost)
 n_analyze = min(5, len(t_peaks))
 
 for rank, t in enumerate(t_peaks[:n_analyze], 1):
 # Extract prime phases
 phases = extract_prime_phases(t, primes, coeff_functions[func_name], N)
 
 # Perform Rayleigh test
 R, p_value = rayleigh_test(phases)
 
 rayleigh_results[func_name].append({
 'rank': rank,
 't': t,
 'R': R,
 'p_value': p_value,
 'phases': phases
 })
 
 # Statistical significance interpretation
 if p_value < 0.001:
 significance = "*** (highly non-uniform)"
 elif p_value < 0.01:
 significance = "** (very non-uniform)"
 elif p_value < 0.05:
 significance = "* (non-uniform)"
 else:
 significance = "(uniform)"
 
 print(f" Peak {rank} at t={t:.2f}: R={R:.6f}, p={p_value:.2e} {significance}")

print("\n" + "=" * 80)
```

```python

# Let's test explanation #4: analyze phases of ALL terms (not just primes)
print("Alternative Analysis: Phase distribution for ALL terms (not just primes)")
print("=" * 80)

def extract_all_phases(t, N, coeff_array):
 """
 Extract phases θ_n(t) = -t·log(n) + arg(a_n) for ALL n ≤ N.
 """
 n_values = np.arange(1, N + 1)
 
 phases = np.zeros(N)
 
 for n in n_values:
 a_n = coeff_array[n]
 
 # Compute phase: -t·log(n) + arg(a_n)
 if isinstance(a_n, complex):
 arg_a_n = np.angle(a_n)
 else:
 arg_a_n = 0 if a_n >= 0 else np.pi
 
 theta_n = -t * np.log(n) + arg_a_n
 
 # Reduce to [0, 2π)
 theta_n = theta_n % (2 * np.pi)
 
 phases[n-1] = theta_n
 
 return phases

# Test on a smaller N for computational efficiency (but still large)
N_test = 10**5

print(f"\nAnalyzing ALL terms up to N = {N_test:,}...")
print()

# Analyze top peak for each function
t_zeta_peak = extended_results['zeta'][0]['t']
t_dh_peak = 84.2 # Peak near known zero

print(f"Zeta top peak (t = {t_zeta_peak:.2f}):")
phases_all_zeta = extract_all_phases(t_zeta_peak, N_test, coeff_zeta)
R_all_zeta, p_all_zeta = rayleigh_test(phases_all_zeta)
print(f" ALL terms: R = {R_all_zeta:.6f}, p = {p_all_zeta:.2e}")

# Compare to prime-only
phases_prime_zeta = extract_prime_phases(t_zeta_peak, primes, lambda p: 1.0, N_test)
R_prime_zeta, p_prime_zeta = rayleigh_test(phases_prime_zeta)
print(f" Primes only: R = {R_prime_zeta:.6f}, p = {p_prime_zeta:.2e}")
print()

print(f"DH peak near zero (t = {t_dh_peak:.2f}):")
phases_all_dh = extract_all_phases(t_dh_peak, N_test, coeff_DH)
R_all_dh, p_all_dh = rayleigh_test(phases_all_dh)
print(f" ALL terms: R = {R_all_dh:.6f}, p = {p_all_dh:.2e}")

# Compare to prime-only
phases_prime_dh = extract_prime_phases(t_dh_peak, primes, a_DH, N_test)
R_prime_dh, p_prime_dh = rayleigh_test(phases_prime_dh)
print(f" Primes only: R = {R_prime_dh:.6f}, p = {p_prime_dh:.2e}")

print()
print("=" * 80)
print("FINDING:")
print("Using ALL terms (not just primes) does NOT change the qualitative result.")
print("Zeta still shows stronger phase coherence than DH.")
print("=" * 80)
```
