```python

# Perfect! The research-program-v3.pdf contains the precise historical definition on page 5 (Section 3.1).
# The Davenport-Heilbronn function is defined as:
#
# L_DH(s) = [(1-iκ)/2] L(s,χ) + [(1+iκ)/2] L(s,χ̄)
#
# where χ is the primitive complex character modulo 5 of order 4:
# χ(1) = 1, χ(2) = i, χ(3) = -i, χ(4) = -1, χ(0) = 0
#
# and κ = (√5 - 1) / (2√(5(√5-1)))
#
# The four known off-line zeros from Spira (1994) and Balanzario-Sánchez-Ortiz (2007):
# (σ, t) = (0.8085, 85.70), (0.6508, 114.16), (0.5744, 166.48), (0.7243, 176.70)

import numpy as np

# Calculate κ
sqrt5 = np.sqrt(5)
kappa = (sqrt5 - 1) / (2 * np.sqrt(5 * (sqrt5 - 1)))
theta = np.arctan(kappa)

print("Historical Davenport-Heilbronn Function Parameters:")
print("=" * 80)
print(f"κ = {kappa:.10f}")
print(f"θ = arctan(κ) = {theta:.10f} radians = {np.degrees(theta):.6f} degrees")
print(f"\nCoefficients:")
print(f" L_DH(s) = [(1-iκ)/2] L(s,χ) + [(1+iκ)/2] L(s,χ̄)")
print(f" where (1-iκ)/2 = {(1-1j*kappa)/2}")
print(f" and (1+iκ)/2 = {(1+1j*kappa)/2}")
print("\nKnown off-line zeros:")
known_zeros = [
 (0.8085, 85.70),
 (0.6508, 114.16),
 (0.5744, 166.48),
 (0.7243, 176.70)
]
for i, (sigma, t) in enumerate(known_zeros, 1):
 print(f" ρ_{i} = {sigma} + i·{t}")
```

```python

# Now let me implement the Dirichlet coefficient function for L_DH
# From equation (2) in the research program:
# a_n = [(1-iκ)/2] χ(n) + [(1+iκ)/2] χ̄(n)

def chi_mod5(n):
 """Primitive complex character modulo 5 of order 4
 χ(1) = 1, χ(2) = i, χ(3) = -i, χ(4) = -1, χ(0) = 0
 """
 n_mod = n % 5
 if n_mod == 0:
 return 0
 elif n_mod == 1:
 return 1
 elif n_mod == 2:
 return 1j
 elif n_mod == 3:
 return -1j
 else: # n_mod == 4
 return -1

def a_DH_historical(n, kappa):
 """Historical Davenport-Heilbronn coefficients
 a_n = [(1-iκ)/2] χ(n) + [(1+iκ)/2] χ̄(n)
 """
 chi_n = chi_mod5(n)
 chi_bar_n = np.conj(chi_n)
 
 coeff1 = (1 - 1j*kappa) / 2
 coeff2 = (1 + 1j*kappa) / 2
 
 return coeff1 * chi_n + coeff2 * chi_bar_n

# Test the coefficients
print("Davenport-Heilbronn coefficients a_n:")
print("=" * 80)
print("n χ(n) χ̄(n) a_n (historical)")
print("-" * 80)
for n in range(1, 21):
 chi_n = chi_mod5(n)
 chi_bar_n = np.conj(chi_n)
 a_n = a_DH_historical(n, kappa)
 print(f"{n:2d} {chi_n:>6} {chi_bar_n:>6} {a_n.real:8.5f} + {a_n.imag:8.5f}i")

# Check that coefficients are real when n ≡ 1, 4 (mod 5)
print("\n" + "=" * 80)
print("Verification: a_n should be REAL when n ≡ 1, 4 (mod 5)")
print(" a_n should be COMPLEX when n ≡ 2, 3 (mod 5)")
for n in range(1, 11):
 a_n = a_DH_historical(n, kappa)
 n_mod = n % 5
 is_real = abs(a_n.imag) < 1e-10
 print(f"n={n:2d} (≡{n_mod} mod 5): a_n = {a_n:20}, real? {is_real}")
```
